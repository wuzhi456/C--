```python
#æ•°æ®é¢„å¤„ç†
import pandas as pd
import numpy as np
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# ==================== é…ç½®åŒº ====================
DATA_DIR = './'
OUTPUT_PATH = DATA_DIR + 'panel_data_clean_v2.csv'

# èµ›å­£å¯åŠ¨æ—¶é—´æŸ¥æ‰¾è¡¨
SEASON_START_DATES = {
    1: '2005-06-01', 2: '2006-01-05', 3: '2006-09-12', 4: '2007-03-19',
    5: '2007-09-24', 6: '2008-03-17', 7: '2008-09-22', 8: '2009-03-09',
    9: '2009-09-21', 10: '2010-03-22', 11: '2010-09-20', 12: '2011-03-21',
    13: '2011-09-19', 14: '2012-03-19', 15: '2012-09-24', 16: '2013-03-18',
    17: '2013-09-16', 18: '2014-03-17', 19: '2014-09-15', 20: '2015-03-16',
    21: '2015-09-14', 22: '2016-03-21', 23: '2016-09-12', 24: '2017-03-20',
    25: '2017-09-18', 26: '2018-04-30', 27: '2018-09-24', 28: '2019-09-16',
    29: '2020-09-14', 30: '2021-09-20', 31: '2022-09-19', 32: '2023-09-26',
    33: '2024-09-17', 34: '2025-03-18',
}

# åå­—æ‰‹åŠ¨æ˜ å°„è¡¨
NAME_MANUAL_MAP = {
    'aj mclean': 'aj mclean',
    'a j mclean': 'aj mclean',
    'dj tanner': 'candace cameron bure',
    'the situation': 'mike sorrentino',
    'vanilla ice': 'robert van winkle',
    'mr t': 'mr t',
}

def get_season_start_date(season):
    if season in SEASON_START_DATES:
        return pd.to_datetime(SEASON_START_DATES[season])
    else:
        return pd.Timestamp('2005-01-01') + pd.DateOffset(months=(season - 1) * 6)

print("=" * 80)
print("ğŸš€ DWTS æ•°æ®æ¸…æ´—ç®¡é“ v2.3 - ç»ˆæä¿®å¤ç‰ˆï¼ˆPandas 2.x å…¼å®¹ï¼‰")
print("=" * 80)

# ==================== æ•°æ®åŠ è½½ ====================
df_main = pd.read_excel(DATA_DIR + '2026_MCM_Problem_C_Data_plus_follers.xlsx')
df_trends = pd.read_csv(DATA_DIR + 'dwts_historical_trends.csv')
df_weekly = pd.read_csv(DATA_DIR + 'dwts_weekly_details.csv')

df_main.columns = df_main.columns.str.strip()
print(f"âœ“ æ•°æ®åŠ è½½å®Œæˆ | ä¸»è¡¨: {len(df_main)} | å‘¨è¡¨: {len(df_weekly)} | è¶‹åŠ¿: {len(df_trends)}")

# ==================== ç¬¬ä¸€æ­¥ï¼šæ„å»ºåˆ†æ•°é¢æ¿ï¼ˆå®½è½¬é•¿ï¼‰====================
print("\n[æ­¥éª¤1] æ„å»ºåˆ†æ•°é¢æ¿...")

panel_data = []

for idx, row in df_main.iterrows():
    for week in range(1, 12):
        week_cols = [f'week{week}_judge{j}_score' for j in range(1, 5)]
        existing_cols = [col for col in week_cols if col in df_main.columns]
        
        if not existing_cols:
            continue
            
        scores = [pd.to_numeric(row[col], errors='coerce') for col in existing_cols]
        valid_scores = [s for s in scores if pd.notna(s) and s > 0]
        
        if len(valid_scores) == 0:
            continue
        
        panel_data.append({
            'celebrity_name': str(row['celebrity_name']).strip(),
            'ballroom_partner': row['ballroom_partner'],
            'celebrity_industry': row['celebrity_industry'],
            'celebrity_age': pd.to_numeric(row['celebrity_age_during_season'], errors='coerce'),
            'celebrity_instagram_followers': pd.to_numeric(row.get('celebrity_instagram_followers'), errors='coerce'),
            'celebrity_twitter_followers': pd.to_numeric(row.get('celebrity_twitter_followers'), errors='coerce'),
            'celebrity_tiktok_followers': pd.to_numeric(row.get('celebrity_tiktok_followers'), errors='coerce'),
            'celebrity_total_followers_wikidata': pd.to_numeric(row.get('celebrity_total_followers_wikidata'), errors='coerce'),
            'partner_instagram_followers': pd.to_numeric(row.get('partner_instagram_followers'), errors='coerce'),
            'partner_twitter_followers': pd.to_numeric(row.get('partner_twitter_followers'), errors='coerce'),
            'partner_tiktok_followers': pd.to_numeric(row.get('partner_tiktok_followers'), errors='coerce'),
            'partner_total_followers_wikidata': pd.to_numeric(row.get('partner_total_followers_wikidata'), errors='coerce'),
            'season': int(row['season']) if pd.notna(row['season']) else np.nan,
            'week': week,
            'judge1_score': scores[0] if len(scores) > 0 else np.nan,
            'judge2_score': scores[1] if len(scores) > 1 else np.nan,
            'judge3_score': scores[2] if len(scores) > 2 else np.nan,
            'judge4_score': scores[3] if len(scores) > 3 else np.nan,
            'weekly_total': sum(valid_scores),
            'weekly_avg': np.mean(valid_scores),
            'num_judges': len(valid_scores)
        })

df_score = pd.DataFrame(panel_data)
print(f"âœ“ åˆ†æ•°é¢æ¿æ„å»ºå®Œæˆ: {len(df_score)} è¡Œ")

# ==================== ç¬¬äºŒæ­¥ï¼šæ¸…æ´—ç»“æœè¡¨ ====================
print("\n[æ­¥éª¤2] æ¸…æ´—ç»“æœè¡¨...")

df_weekly['Season'] = pd.to_numeric(df_weekly['Season'], errors='coerce')
df_weekly['Week'] = pd.to_numeric(df_weekly['Week'], errors='coerce')
df_weekly = df_weekly.dropna(subset=['Season', 'Week'])
df_weekly['Season'] = df_weekly['Season'].astype(int)
df_weekly['Week'] = df_weekly['Week'].astype(int)

df_weekly['Celebrity'] = df_weekly['Celebrity'].fillna('')
df_weekly['celebrity_raw'] = df_weekly['Celebrity'].str.strip()
df_weekly['celebrity_first'] = df_weekly['celebrity_raw'].str.split('&').str[0].str.strip()

df_weekly['is_eliminated'] = df_weekly['Weekly_Bottom_Two_Status'].str.contains('Eliminated', case=False, na=False).astype(int)
df_weekly['is_bottom_two'] = df_weekly['Weekly_Bottom_Two_Status'].str.contains('Bottom', case=False, na=False).astype(int)
df_weekly['is_safe'] = df_weekly['Weekly_Bottom_Two_Status'].str.contains('Safe', case=False, na=False).astype(int)
df_weekly['is_winner'] = df_weekly['Weekly_Bottom_Two_Status'].str.contains('Winner', case=False, na=False).astype(int)

df_weekly.loc[df_weekly['is_eliminated'] == 1, 'is_bottom_two'] = 1

df_status = df_weekly[['Season', 'Week', 'celebrity_first', 'celebrity_raw', 'Dance_Style', 'Running_Order',
                       'is_eliminated', 'is_bottom_two', 'is_safe', 'is_winner']].copy()
df_status.columns = ['season', 'week', 'celebrity_name', 'celebrity_full', 'dance_style', 'running_order',
                     'is_eliminated', 'is_bottom_two', 'is_safe', 'is_winner']

print(f"âœ“ ç»“æœè¡¨æ¸…æ´—å®Œæˆ: {len(df_status)} è¡Œ")

# ==================== ç¬¬ä¸‰æ­¥ï¼šæ¨¡ç³ŠåŒ¹é… ====================
print("\n[æ­¥éª¤3] æ—¶ç©ºåæ ‡é”å®š + æ¨¡ç³ŠåŒ¹é…...")

def normalize_name(name):
    """å¢å¼ºç‰ˆåå­—æ ‡å‡†åŒ–"""
    if pd.isna(name):
        return ''
    name = str(name).lower()
    name = name.replace("'", "").replace("-", " ").replace(".", "").replace(",", "")
    name = name.replace("&", " ").replace("  ", " ")
    name = ' '.join(name.split())
    
    if name in NAME_MANUAL_MAP:
        return NAME_MANUAL_MAP[name]
    
    return name

df_score['name_norm'] = df_score['celebrity_name'].apply(normalize_name)
df_status['name_norm'] = df_status['celebrity_name'].apply(normalize_name)
df_status['name_full_norm'] = df_status['celebrity_full'].apply(normalize_name)

matched_records = []
unmatched_score = []

for (season, week), score_group in df_score.groupby(['season', 'week']):
    status_group = df_status[(df_status['season'] == season) & (df_status['week'] == week)]
    
    if len(status_group) == 0:
        unmatched_score.extend(score_group.to_dict('records'))
        continue
    
    for _, score_row in score_group.iterrows():
        score_name = score_row['name_norm']
        matched = False
        
        for _, status_row in status_group.iterrows():
            status_name = status_row['name_norm']
            status_full = status_row['name_full_norm']
            
            if (status_name and status_name in score_name) or \
               (score_name and score_name in status_name) or \
               (status_full and (status_full in score_name or score_name in status_full)):
                matched_records.append({**score_row.to_dict(), **status_row.to_dict()})
                matched = True
                break
            
            score_last = score_name.split()[-1] if score_name else ''
            status_last = status_name.split()[-1] if status_name else ''
            if score_last and status_last and len(score_last) > 2 and score_last == status_last:
                matched_records.append({**score_row.to_dict(), **status_row.to_dict()})
                matched = True
                break
        
        if not matched:
            unmatched_score.append(score_row.to_dict())

df_panel = pd.DataFrame(matched_records)

print(f"âœ“ åŒ¹é…æˆåŠŸ: {len(df_panel)} è¡Œ ({len(df_panel)/len(df_score)*100:.1f}%)")
print(f"âœ— åŒ¹é…å¤±è´¥: {len(unmatched_score)} è¡Œ")

# ==================== ç¬¬å››æ­¥ï¼šå»é‡ ====================
print("\n[æ­¥éª¤4] å»é‡...")

df_panel = df_panel.drop_duplicates(subset=['season', 'week', 'celebrity_name', 'weekly_total'])

df_panel['status_priority'] = (
    df_panel['is_eliminated'] * 4 + 
    df_panel['is_winner'] * 3 + 
    df_panel['is_bottom_two'] * 2 + 
    df_panel['is_safe'] * 1
)
df_panel = df_panel.sort_values('status_priority', ascending=False)
df_panel = df_panel.drop_duplicates(subset=['season', 'week', 'celebrity_name'], keep='first')
df_panel = df_panel.drop('status_priority', axis=1)

print(f"âœ“ å»é‡å®Œæˆ: {len(df_panel)} è¡Œ")

# ==================== ğŸ”§ ä¿®å¤C: ä¿®å¤èˆç§ç¼ºå¤±ï¼ˆPandas 2.x å…¼å®¹ç‰ˆï¼‰====================
print("\n[æ­¥éª¤4.5] ğŸ”§ ä¿®å¤èˆç§ç¼ºå¤±...")

df_panel = df_panel.sort_values(['celebrity_name', 'season', 'week']).reset_index(drop=True)

follower_cols = [
    'celebrity_instagram_followers', 'celebrity_twitter_followers', 'celebrity_tiktok_followers',
    'celebrity_total_followers_wikidata', 'partner_instagram_followers', 'partner_twitter_followers',
    'partner_tiktok_followers', 'partner_total_followers_wikidata'
]
df_panel[follower_cols] = df_panel[follower_cols].apply(pd.to_numeric, errors='coerce')
celebrity_social_sum = df_panel[
    ['celebrity_instagram_followers', 'celebrity_twitter_followers', 'celebrity_tiktok_followers']
].sum(axis=1, min_count=1)
partner_social_sum = df_panel[
    ['partner_instagram_followers', 'partner_twitter_followers', 'partner_tiktok_followers']
].sum(axis=1, min_count=1)
df_panel['celebrity_total_followers'] = df_panel['celebrity_total_followers_wikidata'].where(
    df_panel['celebrity_total_followers_wikidata'].notna(), celebrity_social_sum
)
df_panel['partner_total_followers'] = df_panel['partner_total_followers_wikidata'].where(
    df_panel['partner_total_followers_wikidata'].notna(), partner_social_sum
)
EMPTY_DATASET_DEFAULT = 0
follower_median = pd.concat([
    df_panel['celebrity_total_followers'],
    df_panel['partner_total_followers']
], ignore_index=True).dropna().median()
follower_median = follower_median if pd.notna(follower_median) else EMPTY_DATASET_DEFAULT
df_panel['celebrity_total_followers'] = df_panel['celebrity_total_followers'].fillna(follower_median)
df_panel['partner_total_followers'] = df_panel['partner_total_followers'].fillna(follower_median)
df_panel['total_followers_log'] = np.log1p(df_panel['celebrity_total_followers'] + df_panel['partner_total_followers'])

# ğŸ”§ ä¿®å¤ï¼šä½¿ç”¨ transform + ffill/bfillï¼ˆPandas 2.x å…¼å®¹ï¼‰
df_panel['dance_style'] = df_panel.groupby(['celebrity_name', 'season'])['dance_style'].transform(
    lambda x: x.ffill().bfill()
)
df_panel['dance_style'] = df_panel['dance_style'].fillna('Unknown')

missing_dance = df_panel['dance_style'].isna().sum()
print(f"âœ“ èˆç§ç¼ºå¤±ä¿®å¤å®Œæˆ (å‰©ä½™ç¼ºå¤±: {missing_dance})")

# ==================== ğŸ”§ ä¿®å¤B: ä¿®æ­£åˆ†æ•°ç‡è®¡ç®— ====================
print("\n[æ­¥éª¤5] ğŸ”§ æ ‡å‡†åŒ–åˆ†æ•°ï¼ˆä¿®æ­£ç‰ˆï¼‰...")

df_panel = df_panel.sort_values(['season', 'week', 'celebrity_name']).reset_index(drop=True)

# æ£€æµ‹å¥–é‡‘åˆ†ï¼ˆBonusï¼‰
def extract_bonus(row):
    scores = [row['judge1_score'], row['judge2_score'], row['judge3_score'], row['judge4_score']]
    scores = [s for s in scores if pd.notna(s)]
    bonus = sum([max(0, s - 10) for s in scores])
    return bonus

df_panel['bonus_points'] = df_panel.apply(extract_bonus, axis=1)

# é‡æ–°è®¡ç®—ç†è®ºæœ€é«˜åˆ†
df_panel['theoretical_max'] = df_panel['num_judges'] * 10

# ä¿®æ­£ï¼šscore_pct = (weekly_total - bonus) / theoretical_max
df_panel['score_pct'] = (df_panel['weekly_total'] - df_panel['bonus_points']) / df_panel['theoretical_max']
df_panel['score_pct'] = df_panel['score_pct'].clip(0, 1)

# ç›¸å¯¹åˆ†æ•°å æ¯”
week_totals = df_panel.groupby(['season', 'week'])['weekly_total'].transform('sum')
df_panel['relative_score_pct'] = df_panel['weekly_total'] / week_totals

# æ’å
df_panel['weekly_rank'] = df_panel.groupby(['season', 'week'])['weekly_total'].rank(ascending=False, method='min')

print(f"âœ“ æ ‡å‡†åŒ–ç‰¹å¾ç”Ÿæˆå®Œæˆ")
print(f"  æ£€æµ‹åˆ°å¥–é‡‘åˆ†: {(df_panel['bonus_points'] > 0).sum()} æ¬¡")
print(f"  score_pct ä¿®æ­£åèŒƒå›´: [{df_panel['score_pct'].min():.3f}, {df_panel['score_pct'].max():.3f}]")

# ==================== ç¬¬å…­æ­¥ï¼šåŠ¨æ€ç‰¹å¾ ====================
print("\n[æ­¥éª¤6] ç”ŸæˆåŠ¨æ€ç‰¹å¾...")

df_panel = df_panel.sort_values(['celebrity_name', 'season', 'week']).reset_index(drop=True)

df_panel['cumulative_avg'] = df_panel.groupby(['celebrity_name', 'season'])['weekly_avg'].transform(
    lambda x: x.expanding().mean()
)
df_panel['cumulative_total'] = df_panel.groupby(['celebrity_name', 'season'])['weekly_total'].cumsum()
df_panel['score_volatility'] = df_panel.groupby(['celebrity_name', 'season'])['weekly_avg'].transform(
    lambda x: x.expanding().std()
).fillna(0)
df_panel['score_deviation'] = df_panel['weekly_avg'] - df_panel['cumulative_avg']
df_panel['cumulative_rank'] = df_panel.groupby(['season', 'week'])['cumulative_total'].rank(ascending=False, method='min')
df_panel['weeks_survived'] = df_panel.groupby(['celebrity_name', 'season']).cumcount() + 1

print(f"âœ“ åŠ¨æ€ç‰¹å¾ç”Ÿæˆå®Œæˆ")

# ==================== ç¬¬ä¸ƒæ­¥ï¼šæ»åç‰¹å¾ ====================
print("\n[æ­¥éª¤7] æ„é€ æ»åç‰¹å¾...")

df_panel = df_panel.sort_values(['celebrity_name', 'season', 'week']).reset_index(drop=True)
df_panel['was_bottom_last_week'] = df_panel.groupby(['celebrity_name', 'season'])['is_bottom_two'].shift(1).fillna(0).astype(int)
df_panel['was_eliminated_last_week'] = df_panel.groupby(['celebrity_name', 'season'])['is_eliminated'].shift(1).fillna(0).astype(int)
df_panel['score_change'] = df_panel.groupby(['celebrity_name', 'season'])['weekly_avg'].diff().fillna(0)

print(f"âœ“ æ»åç‰¹å¾ç”Ÿæˆå®Œæˆ")

# ==================== ğŸ”§ ä¿®å¤A: å¼ºåŒ–æµè¡Œåº¦åŒ¹é… ====================
print("\n[æ­¥éª¤8] ğŸ”§ åˆå¹¶æœç´¢çƒ­åº¦ï¼ˆå¼ºåŒ–åŒ¹é…ç‰ˆï¼‰...")

df_trends['date'] = pd.to_datetime(df_trends['date'])
df_trends['search_index'] = pd.to_numeric(df_trends['search_index'], errors='coerce').fillna(0)
df_trends['celebrity_name_norm'] = df_trends['celebrity_name'].apply(normalize_name)

# è¯Šæ–­ï¼šæ‰“å°äº¤é›†ä¿¡æ¯
panel_names = set(df_panel['name_norm'].unique())
trend_names = set(df_trends['celebrity_name_norm'].unique())
exact_match = panel_names & trend_names

print(f"\nğŸ“Š åå­—åŒ¹é…è¯Šæ–­:")
print(f"  é¢æ¿ä¸­å”¯ä¸€é€‰æ‰‹: {len(panel_names)}")
print(f"  è¶‹åŠ¿ä¸­å”¯ä¸€é€‰æ‰‹: {len(trend_names)}")
print(f"  ç²¾ç¡®åŒ¹é…: {len(exact_match)} ({len(exact_match)/len(panel_names)*100:.1f}%)")

# æ¨¡ç³ŠåŒ¹é…å‡½æ•°
def fuzzy_match_celebrity(panel_name, trend_names_set, trend_tokens_map):
    if not panel_name:
        return None
    if panel_name in trend_names_set:
        return panel_name
    
    substring_matches = [
        trend_name for trend_name in trend_names_set
        if panel_name in trend_name or trend_name in panel_name
    ]
    if substring_matches:
        return min(substring_matches, key=len)
    
    panel_tokens = panel_name.split()
    if panel_tokens:
        panel_last = panel_tokens[-1]
        if len(panel_last) > 2:
            last_matches = [
                trend_name for trend_name in trend_names_set
                if (trend_name.split()[-1] if trend_name else '') == panel_last
            ]
            if last_matches:
                return min(last_matches, key=len)
        
        panel_token_set = set(panel_tokens)
        best_match = None
        best_overlap = 0
        for trend_name, token_set in trend_tokens_map.items():
            overlap = len(panel_token_set & token_set)
            if overlap > best_overlap and overlap >= 2:
                best_match = trend_name
                best_overlap = overlap
        if best_match:
            return best_match
    
    return None

# é¢„å…ˆæ„å»ºåå­—æ˜ å°„ï¼Œé¿å…é‡å¤åŒ¹é…
trend_tokens_map = {name: set(name.split()) for name in trend_names}
trend_name_map = {
    panel_name: fuzzy_match_celebrity(panel_name, trend_names, trend_tokens_map)
    for panel_name in panel_names
}

# è®¡ç®—æµè¡Œåº¦ç‰¹å¾
popularity_features = []
match_success = 0
match_fail = 0

for _, row in df_panel.iterrows():
    celebrity = row['name_norm']
    season = row['season']
    
    cutoff_date = get_season_start_date(season)
    
    matched_name = trend_name_map.get(celebrity)
    
    if matched_name:
        trend_subset = df_trends[
            (df_trends['celebrity_name_norm'] == matched_name) & 
            (df_trends['date'] < cutoff_date)
        ]
        match_success += 1
    else:
        trend_subset = pd.DataFrame()
        match_fail += 1
    
    if len(trend_subset) > 0:
        search_mean = trend_subset['search_index'].mean()
        search_max = trend_subset['search_index'].max()
        search_std = trend_subset['search_index'].std()
        search_recent_3m = trend_subset.tail(3)['search_index'].mean()
    else:
        search_mean = 0
        search_max = 0
        search_std = 0
        search_recent_3m = 0
    
    popularity_features.append({
        'search_mean_pre_season': search_mean,
        'search_max_pre_season': search_max,
        'search_std_pre_season': search_std if pd.notna(search_std) else 0,
        'search_recent_3m': search_recent_3m,
        'search_mean_log': np.log1p(search_mean),
        'search_max_log': np.log1p(search_max),
        'search_std_log': np.log1p(search_std if pd.notna(search_std) else 0),
        'search_recent_3m_log': np.log1p(search_recent_3m),
    })

df_popularity = pd.DataFrame(popularity_features)
df_panel = pd.concat([df_panel.reset_index(drop=True), df_popularity], axis=1)

print(f"\nâœ“ æµè¡Œåº¦ç‰¹å¾åˆå¹¶å®Œæˆ")
print(f"  æ¨¡ç³ŠåŒ¹é…æˆåŠŸ: {match_success} ({match_success/len(df_panel)*100:.1f}%)")
print(f"  æ¨¡ç³ŠåŒ¹é…å¤±è´¥: {match_fail} ({match_fail/len(df_panel)*100:.1f}%)")

# ==================== ç¬¬ä¹æ­¥ï¼šèŒä¸šç¼–ç  ====================
print("\n[æ­¥éª¤9] èŒä¸šç¼–ç ...")

industry_mapping = {
    'Actor/Actress': 1,
    'Athlete': 2,
    'Singer/Rapper': 3,
    'TV Personality': 4,
    'Model': 5,
    'News Anchor': 6,
    'Sports Broadcaster': 7,
}
df_panel['industry_code'] = df_panel['celebrity_industry'].map(industry_mapping).fillna(0).astype(int)

# ==================== ç¬¬åæ­¥ï¼šæœ€ç»ˆæ•´ç† ====================
print("\n[æ­¥éª¤10] æœ€ç»ˆæ•´ç†...")

key_columns = [
    'season', 'week', 'celebrity_name', 'ballroom_partner',
    'celebrity_age', 'celebrity_industry', 'industry_code',
    'judge1_score', 'judge2_score', 'judge3_score', 'judge4_score',
    'weekly_total', 'weekly_avg', 'num_judges',
    'bonus_points',
    'score_pct', 'relative_score_pct',
    'cumulative_avg', 'cumulative_total', 'score_volatility', 'score_deviation',
    'weekly_rank', 'cumulative_rank', 'weeks_survived',
    'was_bottom_last_week', 'was_eliminated_last_week', 'score_change',
    'dance_style', 'running_order',
    'is_eliminated', 'is_bottom_two', 'is_safe', 'is_winner',
    'celebrity_total_followers', 'partner_total_followers', 'total_followers_log',
    'search_mean_pre_season', 'search_max_pre_season', 'search_std_pre_season', 'search_recent_3m',
    'search_mean_log', 'search_max_log', 'search_std_log', 'search_recent_3m_log'
]

df_panel_final = df_panel[key_columns].copy()
df_panel_final = df_panel_final.sort_values(['season', 'week', 'celebrity_name']).reset_index(drop=True)

df_panel_final.to_csv(OUTPUT_PATH, index=False)

# ==================== éªŒè¯æŠ¥å‘Š ====================
print("\n" + "=" * 80)
print("âœ… æ•°æ®æ¸…æ´—å®Œæˆï¼")
print("=" * 80)

print(f"\nğŸ“Š æ•°æ®å½¢çŠ¶: {df_panel_final.shape}")

print(f"\nğŸ¯ ç›®æ ‡å˜é‡åˆ†å¸ƒ:")
print(f"  æ·˜æ±° (Eliminated): {df_panel_final['is_eliminated'].sum()}")
print(f"  å±é™© (Bottom Two): {df_panel_final['is_bottom_two'].sum()}")
print(f"  å®‰å…¨ (Safe): {df_panel_final['is_safe'].sum()}")
print(f"  å† å†› (Winner): {df_panel_final['is_winner'].sum()}")

print(f"\nğŸ“ˆ æ•°æ®è¦†ç›–:")
print(f"  èµ›å­£æ•°: {df_panel_final['season'].nunique()}")
print(f"  é€‰æ‰‹æ•°: {df_panel_final['celebrity_name'].nunique()}")
print(f"  å¹³å‡å‚èµ›å‘¨æ•°: {df_panel_final.groupby('celebrity_name')['week'].count().mean():.1f}")

print(f"\nğŸ” ä¿®å¤éªŒè¯:")
print(f"  ğŸ”§ score_pct èŒƒå›´: [{df_panel_final['score_pct'].min():.3f}, {df_panel_final['score_pct'].max():.3f}] âœ“")
print(f"  ğŸ”§ relative_score_pct èŒƒå›´: [{df_panel_final['relative_score_pct'].min():.3f}, {df_panel_final['relative_score_pct'].max():.3f}]")
print(f"  ğŸ”§ bonus_points æ£€æµ‹: {(df_panel_final['bonus_points'] > 0).sum()} æ¬¡")
print(f"  ğŸ”§ search_mean_log éé›¶ç‡: {(df_panel_final['search_mean_log'] > 0).mean():.2%} â­")
print(f"  ğŸ”§ was_bottom_last_week æ¿€æ´»ç‡: {df_panel_final['was_bottom_last_week'].mean():.2%}")

print(f"\nâš ï¸  ç¼ºå¤±å€¼æ£€æŸ¥:")
missing = df_panel_final.isnull().sum()
if missing.sum() > 0:
    print(missing[missing > 0])
else:
    print("  âœ“ æ— ç¼ºå¤±å€¼")

print(f"\nğŸ’¾ è¾“å‡ºæ–‡ä»¶: {OUTPUT_PATH}")

inconsistent = df_panel_final[(df_panel_final['is_eliminated'] == 1) & (df_panel_final['is_bottom_two'] == 0)]
if len(inconsistent) > 0:
    print(f"\nâŒ è­¦å‘Šï¼šå‘ç° {len(inconsistent)} æ¡é€»è¾‘ä¸ä¸€è‡´è®°å½•")
else:
    print(f"\nâœ… é€»è¾‘ä¸€è‡´æ€§éªŒè¯é€šè¿‡")

print("\n" + "=" * 80)
print("ğŸ¯ ä¸‰å¤§ä¿®å¤éªŒè¯:")
print("=" * 80)
print(f"âœ… æŒ‡ä»¤A (æµè¡Œåº¦åŒ¹é…): {(df_panel_final['search_mean_log'] > 0).mean():.1%} éé›¶ç‡")
print(f"âœ… æŒ‡ä»¤B (åˆ†æ•°ç‡ä¿®æ­£): score_pct âˆˆ [{df_panel_final['score_pct'].min():.3f}, {df_panel_final['score_pct'].max():.3f}]")
print(f"âœ… æŒ‡ä»¤C (èˆç§ç¼ºå¤±): {df_panel_final['dance_style'].isna().sum()} æ¡ç¼ºå¤±")
print("=" * 80)
```

```python
# åŸ Task 1 ä»£ç å·²æ³¨é‡Š
if False:
    #è®­ç»ƒï¼Œ Task 1
    import pandas as pd
    import numpy as np
    import torch
    import torch.nn as nn
    import torch.nn.functional as F
    from torch.utils.data import Dataset, DataLoader
    from sklearn.preprocessing import StandardScaler
    from typing import Dict, Tuple, List
    import warnings
    warnings.filterwarnings('ignore')

    # ==================== é…ç½®åŒº ====================
    DATA_PATH = './panel_data_clean_v2.csv'
    DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

    # ğŸš© æŒ‡ä»¤ 1ï¼šä¿®æ”¹è¶…å‚æ•°é…ç½®
    HYPERPARAMS = {
        'N_max': 20,
        'hidden_dim': 64,
        'lambda_bottom': 0.7,   # â¬†ï¸ è°ƒé«˜ï¼šå¼ºè°ƒå€’æ•°é¢„æµ‹
        'lambda_kl': 0.05,      # â¬‡ï¸ è°ƒä½ï¼šç»™éšå˜é‡æ›´å¤šè‡ªç”±
        'lambda_smooth': 0.1,
        'tau': 0.1,             # â¬†ï¸ è°ƒé«˜åˆå§‹å€¼ï¼šé˜²æ­¢è¿‡åº¦è‡ªä¿¡
        'learning_rate': 5e-4,  # â¬‡ï¸ è°ƒä½ï¼šè®­ç»ƒæ›´ç¨³
        'batch_size': 1,
        'num_epochs': 100,
        'early_stop_patience': 15,
        'num_industries': 8,
        'industry_embed_dim': 8,
    }

    print("=" * 80)
    print("ğŸš€ DWTS éšå˜é‡æ¨æ–­æ¨¡å‹ - ç»ˆæä¿®å¤ç‰ˆ")
    print("=" * 80)
    print(f"è®¾å¤‡: {DEVICE}")

    # ==================== ğŸš© æŒ‡ä»¤ 1ï¼šæ„å»º"èµ›å­£-å‘¨"å¼ é‡åŒ–æ•°æ®ç®¡é“ ====================
    print("\n" + "=" * 80)
    print("ğŸš© æŒ‡ä»¤ 1ï¼šæ„å»ºèµ›å­£-å‘¨å¼ é‡åŒ–æ•°æ®ç®¡é“")
    print("=" * 80)

    class DWTSDataset(Dataset):
        """
        DWTS å¼ é‡åŒ–æ•°æ®é›†
        """
        def __init__(self, df: pd.DataFrame, N_max: int = 20, verbose: bool = False):
            self.N_max = N_max
            self.verbose = verbose
            self.df = df.sort_values(['season', 'week', 'celebrity_name']).reset_index(drop=True)
        
            # ç‰¹å¾åˆ†ç»„
            self.skill_features = ['relative_score_pct', 'score_deviation', 'weekly_rank']
            self.pop_features = ['search_mean_log', 'search_recent_3m_log', 'total_followers_log']
            self.ctx_features = ['was_bottom_last_week', 'weeks_survived', 'running_order']
        
            # æŒ‰ (season, week) åˆ†ç»„
            self.groups = []
            for (season, week), group_df in df.groupby(['season', 'week']):
                # ğŸ”§ ä¿®å¤: æ£€æŸ¥æ˜¯å¦è¶…è¿‡ N_max
                if len(group_df) > N_max:
                    if self.verbose:
                        print(f"è­¦å‘Š: Season {season} Week {week} æœ‰ {len(group_df)} åé€‰æ‰‹ï¼Œè¶…è¿‡ N_max={N_max}")
                    # æˆªæ–­åˆ° N_max
                    group_df = group_df.head(N_max)
                self.groups.append((season, week, group_df))
        
            if self.verbose:
                print(f"âœ“ æ•°æ®é›†æ„å»ºå®Œæˆ:")
                print(f"  - æ€»ç»„æ•°: {len(self.groups)} ä¸ª (season, week) ç»„")
                print(f"  - N_max: {self.N_max}")
    
        def __len__(self):
            return len(self.groups)
    
        def __getitem__(self, idx) -> Dict[str, torch.Tensor]:
            season, week, group_df = self.groups[idx]
            N_actual = len(group_df)
        
            # ğŸ”§ ä¿®å¤: ç¡®ä¿ä¸è¶…è¿‡ N_max
            N_actual = min(N_actual, self.N_max)
            group_df = group_df.head(N_actual)
        
            # æå–ç‰¹å¾çŸ©é˜µ
            X_skill = group_df[self.skill_features].values.astype(np.float32)
            X_pop = group_df[self.pop_features].values.astype(np.float32)
            X_ctx = group_df[self.ctx_features].values.astype(np.float32)
            industry_code = group_df['industry_code'].values.astype(np.int64)
        
            # æå–ç›®æ ‡å˜é‡
            y_elim = group_df['is_eliminated'].values.astype(np.float32)
            y_bottom = group_df['is_bottom_two'].values.astype(np.float32)
        
            # Padding
            mask = np.zeros(self.N_max, dtype=np.float32)
            mask[:N_actual] = 1.0
        
            X_skill_pad = np.zeros((self.N_max, len(self.skill_features)), dtype=np.float32)
            X_pop_pad = np.zeros((self.N_max, len(self.pop_features)), dtype=np.float32)
            X_ctx_pad = np.zeros((self.N_max, len(self.ctx_features)), dtype=np.float32)
            industry_code_pad = np.zeros(self.N_max, dtype=np.int64)
            y_elim_pad = np.zeros(self.N_max, dtype=np.float32)
            y_bottom_pad = np.zeros(self.N_max, dtype=np.float32)
        
            X_skill_pad[:N_actual] = X_skill
            X_pop_pad[:N_actual] = X_pop
            X_ctx_pad[:N_actual] = X_ctx
            industry_code_pad[:N_actual] = industry_code
            y_elim_pad[:N_actual] = y_elim
            y_bottom_pad[:N_actual] = y_bottom
        
            return {
                'X_skill': torch.from_numpy(X_skill_pad),
                'X_pop': torch.from_numpy(X_pop_pad),
                'X_ctx': torch.from_numpy(X_ctx_pad),
                'industry_code': torch.from_numpy(industry_code_pad),
                'y_elim': torch.from_numpy(y_elim_pad),
                'y_bottom': torch.from_numpy(y_bottom_pad),
                'mask': torch.from_numpy(mask),
                'season': season,
                'week': week,
                'N_actual': N_actual,
            }

    # åŠ è½½åŸå§‹æ•°æ®
    print("\n[æ•°æ®åŠ è½½]")
    df_raw = pd.read_csv(DATA_PATH)
    print(f"âœ“ åŸå§‹æ•°æ®: {len(df_raw)} è¡Œ, {len(df_raw.columns)} åˆ—")

    # æ£€æŸ¥æœ€å¤§é€‰æ‰‹æ•°
    max_contestants = df_raw.groupby(['season', 'week']).size().max()
    print(f"âœ“ æ¯å‘¨æœ€å¤§é€‰æ‰‹æ•°: {max_contestants}")

    # ==================== ğŸš© æŒ‡ä»¤ 2ï¼šå®ç°éšå˜é‡æ¨æ–­å¼•æ“ ====================
    print("\n" + "=" * 80)
    print("ğŸš© æŒ‡ä»¤ 2ï¼šå®ç°éšå˜é‡æ¨æ–­å¼•æ“")
    print("=" * 80)

    class LatentInferenceEngine(nn.Module):
        def __init__(self, pop_dim: int, ctx_dim: int, hidden_dim: int = 64, 
                     num_industries: int = 8, industry_embed_dim: int = 8):
            super().__init__()
        
            self.industry_embedding = nn.Embedding(num_industries, industry_embed_dim)
        
            input_dim = pop_dim + industry_embed_dim + ctx_dim + 1
        
            self.prior_mlp = nn.Sequential(
                nn.Linear(input_dim, hidden_dim),
                nn.ReLU(),
                nn.Dropout(0.2),
                nn.Linear(hidden_dim, hidden_dim),
                nn.ReLU(),
                nn.Dropout(0.2),
            )
        
            self.mu_head = nn.Linear(hidden_dim, 1)
            self.logvar_head = nn.Linear(hidden_dim, 1)
        
            self.gamma = nn.Parameter(torch.tensor(0.5))
            self.delta = nn.Parameter(torch.tensor(0.3))
    
        def forward(self, X_pop, X_ctx, industry_code, v_prev, mask):
            B, N = mask.shape
        
            industry_embed = self.industry_embedding(industry_code)
            was_bottom = X_ctx[:, :, 0]
        
            # ğŸš© æŒ‡ä»¤ 2ï¼šå¼ºåˆ¶å‚æ•°éè´Ÿï¼Œç¬¦åˆç¤¾ä¼šå­¦ç›´è§‰
            gamma_clamped = torch.clamp(self.gamma, min=0.0)
            delta_clamped = torch.clamp(self.delta, min=0.0)
            momentum_term = gamma_clamped * v_prev.unsqueeze(-1)
        
            mlp_input = torch.cat([
                X_pop,
                industry_embed,
                X_ctx,
                momentum_term
            ], dim=-1)
        
            h = self.prior_mlp(mlp_input)
            mu_base = self.mu_head(h).squeeze(-1)
            logvar = self.logvar_head(h).squeeze(-1)
        
            mu = mu_base + delta_clamped * was_bottom
        
            if self.training:
                std = torch.exp(0.5 * logvar)
                eps = torch.randn_like(std)
                z = mu + eps * std
            else:
                z = mu
        
            z_masked = z.masked_fill(mask == 0, -1e9)
            v_hat = F.softmax(z_masked, dim=-1)
        
            return v_hat, mu, logvar

    print(f"âœ“ éšå˜é‡æ¨æ–­å¼•æ“å®ç°å®Œæˆ")

    # ==================== ğŸš© æŒ‡ä»¤ 3ï¼šæ„å»ºèµ›å­£æ„ŸçŸ¥èåˆå±‚ ====================
    print("\n" + "=" * 80)
    print("ğŸš© æŒ‡ä»¤ 3ï¼šæ„å»ºèµ›å­£æ„ŸçŸ¥èåˆå±‚")
    print("=" * 80)

    class SeasonAwareFusion(nn.Module):
        def __init__(self, num_seasons: int = 34):
            super().__init__()
            self.alpha_s = nn.Parameter(torch.ones(num_seasons) * 0.5)
            self.g_s = nn.Parameter(torch.ones(num_seasons) * 0.5)
    
        def forward(self, X_skill, v_hat, season_ids, mask, is_test=False):
            B, N = mask.shape
        
            pct_J = X_skill[:, :, 0]
            rank_J = X_skill[:, :, 2]
        
            N_t = mask.sum(dim=1, keepdim=True)
            R_J = (N_t - rank_J) / (N_t - 1 + 1e-8)
            R_J = R_J * mask
        
            v_hat_masked = v_hat.masked_fill(mask == 0, -1e9)
            rank_V = torch.argsort(torch.argsort(v_hat_masked, dim=1, descending=True), dim=1).float()
            rank_V = rank_V + 1
            R_V = (N_t - rank_V) / (N_t - 1 + 1e-8)
            R_V = R_V * mask
        
            # ğŸš© æŒ‡ä»¤ 3ï¼šæµ‹è¯•æ¨¡å¼ä¸‹ï¼Œä½¿ç”¨è®­ç»ƒèµ›å­£å‚æ•°çš„å‡å€¼ï¼ˆå…³é”®ä¿®å¤ï¼ï¼‰
            if is_test:
                # æ’é™¤å½“å‰æµ‹è¯•èµ›å­£ï¼Œå–å…¶ä»–èµ›å­£å‚æ•°å‡å€¼
                alpha_val = torch.sigmoid(self.alpha_s).mean()
                g_val = torch.sigmoid(self.g_s).mean()
                alpha = alpha_val.expand(B)
                g = g_val.expand(B)
            else:
                alpha = torch.sigmoid(self.alpha_s[season_ids])
                g = torch.sigmoid(self.g_s[season_ids])
        
            S_pct = alpha.unsqueeze(1) * pct_J + (1 - alpha.unsqueeze(1)) * v_hat
            S_rank = alpha.unsqueeze(1) * R_J + (1 - alpha.unsqueeze(1)) * R_V
            S_final = g.unsqueeze(1) * S_pct + (1 - g.unsqueeze(1)) * S_rank
            S_final = S_final * mask
        
            return S_final

    print(f"âœ“ èµ›å­£æ„ŸçŸ¥èåˆå±‚å®ç°å®Œæˆ")

    # ==================== å®Œæ•´æ¨¡å‹ ====================
    print("\n" + "=" * 80)
    print("ğŸ”¥ æ„å»ºå®Œæ•´æ¨¡å‹")
    print("=" * 80)

    class DWTSModel(nn.Module):
        def __init__(self, skill_dim, pop_dim, ctx_dim, hidden_dim, num_seasons, tau,
                     num_industries, industry_embed_dim):
            super().__init__()
        
            self.latent_engine = LatentInferenceEngine(
                pop_dim, ctx_dim, hidden_dim, num_industries, industry_embed_dim
            )
            self.fusion_layer = SeasonAwareFusion(num_seasons)
        
            # ğŸš© æŒ‡ä»¤ 4Aï¼šå°† tau å˜æˆå¯å­¦ä¹ å‚æ•°
            self.tau = nn.Parameter(torch.tensor(float(tau)), requires_grad=True)
    
        def forward(self, batch, v_prev=None, is_test=False):
            X_skill = batch['X_skill']
            X_pop = batch['X_pop']
            X_ctx = batch['X_ctx']
            industry_code = batch['industry_code']
            mask = batch['mask']
            season_ids = batch['season'] - 1
            week = batch['week'][0].item()
        
            # ğŸ”§ ä¿®å¤1: ç¬¬ä¸€å‘¨ä½¿ç”¨å‡åŒ€åˆ†å¸ƒ
            if v_prev is None or week == 1:
                N_t = mask.sum(dim=1, keepdim=True)
                v_prev = mask / N_t
        
            v_hat, mu, logvar = self.latent_engine(X_pop, X_ctx, industry_code, v_prev, mask)
        
            # ğŸš© æŒ‡ä»¤ 4Bï¼šä¼ é€’ is_test å‚æ•°
            S_final = self.fusion_layer(X_skill, v_hat, season_ids, mask, is_test=is_test)
        
            S_masked = S_final.masked_fill(mask == 0, 1e9)
        
            # ğŸš© æŒ‡ä»¤ 4Bï¼šé™åˆ¶ tau èŒƒå›´ï¼Œé˜²æ­¢é™¤ä»¥é›¶æˆ–è¿‡åº¦è‡ªä¿¡
            tau_clamped = torch.clamp(self.tau, min=0.05, max=0.5)
            p_elim = F.softmax(-S_masked / tau_clamped, dim=-1)
        
            p_bottom = torch.sigmoid(-S_final)
            p_bottom = p_bottom * mask
        
            return {
                'p_elim': p_elim,
                'p_bottom': p_bottom,
                'v_hat': v_hat,
                'mu': mu,
                'logvar': logvar,
                'S_final': S_final,
            }

    # ==================== ğŸš© æŒ‡ä»¤ 4ï¼šå®šä¹‰å¤åˆæŸå¤±å‡½æ•°ä¸è®­ç»ƒåè®® ====================
    print("\n" + "=" * 80)
    print("ğŸš© æŒ‡ä»¤ 4ï¼šå®šä¹‰å¤åˆæŸå¤±å‡½æ•°ä¸è®­ç»ƒåè®®")
    print("=" * 80)

    class CompositeLoss(nn.Module):
        def __init__(self, lambda_bottom, lambda_kl, lambda_smooth):
            super().__init__()
            self.lambda_bottom = lambda_bottom
            self.lambda_kl = lambda_kl
            self.lambda_smooth = lambda_smooth
    
        def forward(self, outputs, batch, v_hat_prev=None):
            p_elim = outputs['p_elim']
            p_bottom = outputs['p_bottom']
            v_hat = outputs['v_hat']
            mu = outputs['mu']
            logvar = outputs['logvar']
        
            y_elim = batch['y_elim']
            y_bottom = batch['y_bottom']
            mask = batch['mask']
        
            elim_mask = (y_elim == 1) & (mask == 1)
            if elim_mask.sum() > 0:
                L_elim = -torch.log(p_elim[elim_mask] + 1e-10).mean()
            else:
                L_elim = torch.tensor(0.0, device=p_elim.device)
        
            active_mask = mask == 1
            if active_mask.sum() > 0:
                L_bottom = F.binary_cross_entropy(
                    p_bottom[active_mask],
                    y_bottom[active_mask],
                    reduction='mean'
                )
            else:
                L_bottom = torch.tensor(0.0, device=p_bottom.device)
        
            if active_mask.sum() > 0:
                kl_div = -0.5 * (1 + logvar - mu.pow(2) - logvar.exp())
                kl_per_sample = kl_div[active_mask].sum()
                L_KL = kl_per_sample / active_mask.sum()
            else:
                L_KL = torch.tensor(0.0, device=mu.device)
        
            if v_hat_prev is not None and active_mask.sum() > 0:
                smooth_diff = (v_hat - v_hat_prev).pow(2)
                L_smooth = smooth_diff[active_mask].mean()
            else:
                L_smooth = torch.tensor(0.0, device=v_hat.device)
        
            total_loss = (
                L_elim +
                self.lambda_bottom * L_bottom +
                self.lambda_kl * L_KL +
                self.lambda_smooth * L_smooth
            )
        
            loss_dict = {
                'total': total_loss.item(),
                'elim': L_elim.item(),
                'bottom': L_bottom.item(),
                'kl': L_KL.item(),
                'smooth': L_smooth.item(),
            }
        
            return total_loss, loss_dict

    criterion = CompositeLoss(
        lambda_bottom=HYPERPARAMS['lambda_bottom'],
        lambda_kl=HYPERPARAMS['lambda_kl'],
        lambda_smooth=HYPERPARAMS['lambda_smooth']
    )

    print(f"âœ“ æŸå¤±å‡½æ•°å®ä¾‹åŒ–å®Œæˆ")

    # ==================== è¯„ä¼°æŒ‡æ ‡ ====================
    def hit_rate_at_k(p_elim, y_elim, mask, k=2):
        B, N = p_elim.shape
        hits = 0
        total = 0
    
        for i in range(B):
            active_idx = mask[i] == 1
            true_elim_idx = (y_elim[i] == 1) & active_idx
        
            if true_elim_idx.sum() == 0:
                continue
        
            p_elim_active = p_elim[i].clone()
            p_elim_active[~active_idx] = -1
        
            topk_idx = torch.topk(p_elim_active, k=min(k, active_idx.sum().item()))[1]
        
            if true_elim_idx[topk_idx].sum() > 0:
                hits += 1
        
            total += 1
    
        return hits / total if total > 0 else 0.0

    # ==================== è®­ç»ƒä¸è¯„ä¼°å‡½æ•° ====================
    def train_one_epoch(model, dataloader, criterion, optimizer, device):
        model.train()
        total_loss = 0
        loss_components = {'elim': 0, 'bottom': 0, 'kl': 0, 'smooth': 0}
        num_batches = 0
    
        v_hat_cache = {}
    
        for batch in dataloader:
            batch = {k: v.to(device) if isinstance(v, torch.Tensor) else v 
                    for k, v in batch.items()}
        
            season = batch['season'][0].item()
            week = batch['week'][0].item()
        
            key = (season, week - 1)
            v_prev = v_hat_cache.get(key, None)
        
            outputs = model(batch, v_prev=v_prev)
            loss, loss_dict = criterion(outputs, batch, v_hat_prev=v_prev)
        
            optimizer.zero_grad()
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer.step()
        
            v_hat_cache[(season, week)] = outputs['v_hat'].detach()
        
            total_loss += loss.item()
            for k in loss_components:
                loss_components[k] += loss_dict[k]
            num_batches += 1
    
        avg_loss = total_loss / num_batches
        avg_components = {k: v / num_batches for k, v in loss_components.items()}
    
        return avg_loss, avg_components

    def evaluate(model, dataloader, criterion, device):
        model.eval()
        total_loss = 0
        hit_rate_2 = 0
        num_batches = 0
    
        v_hat_cache = {}
    
        with torch.no_grad():
            for batch in dataloader:
                batch = {k: v.to(device) if isinstance(v, torch.Tensor) else v 
                        for k, v in batch.items()}
            
                season = batch['season'][0].item()
                week = batch['week'][0].item()
                key = (season, week - 1)
                v_prev = v_hat_cache.get(key, None)
            
                # ğŸš© æŒ‡ä»¤ 5ï¼šå¼€å¯æµ‹è¯•æ¨¡å¼å¼€å…³
                outputs = model(batch, v_prev=v_prev, is_test=True)
                loss, _ = criterion(outputs, batch, v_hat_prev=v_prev)
            
                hr = hit_rate_at_k(outputs['p_elim'], batch['y_elim'], batch['mask'], k=2)
            
                v_hat_cache[(season, week)] = outputs['v_hat']
            
                total_loss += loss.item()
                hit_rate_2 += hr
                num_batches += 1
    
        avg_loss = total_loss / num_batches
        avg_hit_rate = hit_rate_2 / num_batches
    
        return avg_loss, avg_hit_rate

    # ==================== LOSO äº¤å‰éªŒè¯ ====================
    print("\n" + "=" * 80)
    print("ğŸ”¥ å¼€å§‹ LOSO äº¤å‰éªŒè¯")
    print("=" * 80)

    all_seasons = df_raw['season'].unique()
    test_seasons = sorted(all_seasons)[-5:]

    numerical_features = [
        'relative_score_pct', 'score_deviation', 'weekly_rank',
        'search_mean_log', 'search_recent_3m_log', 'total_followers_log',
        'weeks_survived', 'running_order'
    ]

    loso_results = []

    for test_season in test_seasons:
        print(f"\n{'='*80}")
        print(f"ğŸ¯ æµ‹è¯•èµ›å­£: {test_season}")
        print(f"{'='*80}")
    
        train_df_raw = df_raw[df_raw['season'] != test_season].copy()
        test_df_raw = df_raw[df_raw['season'] == test_season].copy()
    
        scaler = StandardScaler()
        train_df_raw[numerical_features] = scaler.fit_transform(train_df_raw[numerical_features])
        test_df_raw[numerical_features] = scaler.transform(test_df_raw[numerical_features])
    
        train_dataset = DWTSDataset(train_df_raw, N_max=HYPERPARAMS['N_max'], verbose=False)
        test_dataset = DWTSDataset(test_df_raw, N_max=HYPERPARAMS['N_max'], verbose=False)
    
        train_loader = DataLoader(train_dataset, batch_size=1, shuffle=False)
        test_loader = DataLoader(test_dataset, batch_size=1, shuffle=False)
    
        model = DWTSModel(
            skill_dim=3,
            pop_dim=3,
            ctx_dim=3,
            hidden_dim=HYPERPARAMS['hidden_dim'],
            num_seasons=34,
            tau=HYPERPARAMS['tau'],
            num_industries=HYPERPARAMS['num_industries'],
            industry_embed_dim=HYPERPARAMS['industry_embed_dim']
        ).to(DEVICE)
    
        optimizer = torch.optim.Adam(model.parameters(), lr=HYPERPARAMS['learning_rate'])
        scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
            optimizer, mode='min', factor=0.5, patience=5
        )
    
        best_val_loss = float('inf')
        patience_counter = 0
    
        for epoch in range(HYPERPARAMS['num_epochs']):
            train_loss, train_components = train_one_epoch(
                model, train_loader, criterion, optimizer, DEVICE
            )
        
            if (epoch + 1) % 10 == 0:
                print(f"Epoch {epoch+1}: Loss={train_loss:.4f}, "
                      f"elim={train_components['elim']:.4f}, "
                      f"kl={train_components['kl']:.4f}")
            
                if train_loss < best_val_loss:
                    best_val_loss = train_loss
                    patience_counter = 0
                    torch.save(model.state_dict(), f'./best_model_s{test_season}.pth')
                else:
                    patience_counter += 1
            
                if patience_counter >= HYPERPARAMS['early_stop_patience']:
                    print(f"æ—©åœè§¦å‘")
                    break
            
                scheduler.step(train_loss)
    
        model.load_state_dict(torch.load(f'./best_model_s{test_season}.pth'))
        test_loss, test_hit_rate = evaluate(model, test_loader, criterion, DEVICE)
    
        print(f"\næµ‹è¯•ç»“æœ: Loss={test_loss:.4f}, Hit-rate@2={test_hit_rate:.4f}")
    
        loso_results.append({
            'test_season': test_season,
            'test_loss': test_loss,
            'hit_rate_2': test_hit_rate,
        })

    # æ±‡æ€»ç»“æœ
    print(f"\n{'='*80}")
    print(f"ğŸ† LOSO äº¤å‰éªŒè¯ç»“æœæ±‡æ€»")
    print(f"{'='*80}")

    results_df = pd.DataFrame(loso_results)
    print(results_df)
    print(f"\nå¹³å‡ Hit-rate@2: {results_df['hit_rate_2'].mean():.4f}")

    results_df.to_csv('./loso_results.csv', index=False)

    print(f"\nâœ… æ‰€æœ‰ä¿®å¤å®Œæˆï¼")
```

```python
# åŸ Task 1 ä»£ç å·²æ³¨é‡Š
if False:
    #Task 1
    import pandas as pd
    import numpy as np
    import torch
    import torch.nn as nn
    import torch.nn.functional as F
    from torch.utils.data import Dataset, DataLoader
    from sklearn.preprocessing import StandardScaler
    from typing import Dict, Tuple, List
    import warnings
    warnings.filterwarnings('ignore')

    # ==================== é…ç½®åŒº ====================
    DATA_PATH = './panel_data_clean_v2.csv'
    OUTPUT_PATH = './estimated_fan_votes.csv'
    DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

    # è¶…å‚æ•°é…ç½®ï¼ˆä½¿ç”¨éªŒè¯æˆåŠŸçš„é…ç½®ï¼‰
    HYPERPARAMS = {
        'N_max': 20,
        'hidden_dim': 64,
        'lambda_bottom': 0.7,
        'lambda_top2': 0.5,
        'lambda_kl': 0.05,
        'lambda_smooth': 0.1,
        'tau': 0.1,
        'learning_rate': 5e-4,
        'batch_size': 1,
        'num_epochs': 60,
        'num_industries': 8,
        'industry_embed_dim': 8,
    }

    print("=" * 80)
    print("ğŸš€ DWTS éšå˜é‡æå–ç®¡é“ - ç»ˆæä¿®å¤ç‰ˆ")
    print("=" * 80)
    print(f"è®¾å¤‡: {DEVICE}")

    # ==================== æ•°æ®é›†å®šä¹‰ ====================
    class DWTSDataset(Dataset):
        """DWTS å¼ é‡åŒ–æ•°æ®é›†"""
        def __init__(self, df: pd.DataFrame, N_max: int = 20, verbose: bool = False):
            self.N_max = N_max
            self.verbose = verbose
            self.df = df.sort_values(['season', 'week', 'celebrity_name']).reset_index(drop=True)
        
            # ç‰¹å¾åˆ†ç»„
            self.skill_features = ['relative_score_pct', 'score_deviation', 'weekly_rank']
            self.pop_features = ['search_mean_log', 'search_recent_3m_log', 'total_followers_log']
            self.ctx_features = ['was_bottom_last_week', 'weeks_survived', 'running_order']
        
            # æŒ‰ (season, week) åˆ†ç»„
            self.groups = []
            for (season, week), group_df in df.groupby(['season', 'week']):
                if len(group_df) > N_max:
                    if self.verbose:
                        print(f"è­¦å‘Š: Season {season} Week {week} æœ‰ {len(group_df)} åé€‰æ‰‹ï¼Œæˆªæ–­åˆ° {N_max}")
                    group_df = group_df.head(N_max)
                self.groups.append((season, week, group_df))
        
            if self.verbose:
                print(f"âœ“ æ•°æ®é›†æ„å»ºå®Œæˆ: {len(self.groups)} ä¸ª (season, week) ç»„")
    
        def __len__(self):
            return len(self.groups)
    
        def __getitem__(self, idx) -> Dict[str, torch.Tensor]:
            season, week, group_df = self.groups[idx]
            N_actual = min(len(group_df), self.N_max)
            group_df = group_df.head(N_actual)
        
            # æå–ç‰¹å¾çŸ©é˜µ
            X_skill = group_df[self.skill_features].values.astype(np.float32)
            X_pop = group_df[self.pop_features].values.astype(np.float32)
            X_ctx = group_df[self.ctx_features].values.astype(np.float32)
            industry_code = group_df['industry_code'].values.astype(np.int64)
        
            # æå–ç›®æ ‡å˜é‡
            y_elim = group_df['is_eliminated'].values.astype(np.float32)
            y_bottom = group_df['is_bottom_two'].values.astype(np.float32)
        
            # Padding
            mask = np.zeros(self.N_max, dtype=np.float32)
            mask[:N_actual] = 1.0
        
            X_skill_pad = np.zeros((self.N_max, len(self.skill_features)), dtype=np.float32)
            X_pop_pad = np.zeros((self.N_max, len(self.pop_features)), dtype=np.float32)
            X_ctx_pad = np.zeros((self.N_max, len(self.ctx_features)), dtype=np.float32)
            industry_code_pad = np.zeros(self.N_max, dtype=np.int64)
            y_elim_pad = np.zeros(self.N_max, dtype=np.float32)
            y_bottom_pad = np.zeros(self.N_max, dtype=np.float32)
        
            X_skill_pad[:N_actual] = X_skill
            X_pop_pad[:N_actual] = X_pop
            X_ctx_pad[:N_actual] = X_ctx
            industry_code_pad[:N_actual] = industry_code
            y_elim_pad[:N_actual] = y_elim
            y_bottom_pad[:N_actual] = y_bottom
        
            return {
                'X_skill': torch.from_numpy(X_skill_pad),
                'X_pop': torch.from_numpy(X_pop_pad),
                'X_ctx': torch.from_numpy(X_ctx_pad),
                'industry_code': torch.from_numpy(industry_code_pad),
                'y_elim': torch.from_numpy(y_elim_pad),
                'y_bottom': torch.from_numpy(y_bottom_pad),
                'mask': torch.from_numpy(mask),
                'season': season,
                'week': week,
                'N_actual': N_actual,
            }

    # ==================== æ¨¡å‹å®šä¹‰ ====================
    class LatentInferenceEngine(nn.Module):
        """éšå˜é‡æ¨æ–­å¼•æ“"""
        def __init__(self, pop_dim: int, ctx_dim: int, hidden_dim: int = 64, 
                     num_industries: int = 8, industry_embed_dim: int = 8):
            super().__init__()
        
            self.industry_embedding = nn.Embedding(num_industries, industry_embed_dim)
            input_dim = pop_dim + industry_embed_dim + ctx_dim + 1
        
            self.prior_mlp = nn.Sequential(
                nn.Linear(input_dim, hidden_dim),
                nn.ReLU(),
                nn.Dropout(0.2),
                nn.Linear(hidden_dim, hidden_dim),
                nn.ReLU(),
                nn.Dropout(0.2),
            )
        
            self.mu_head = nn.Linear(hidden_dim, 1)
            self.logvar_head = nn.Linear(hidden_dim, 1)
        
            self.gamma = nn.Parameter(torch.tensor(0.5))
            self.delta = nn.Parameter(torch.tensor(0.3))
    
        def forward(self, X_pop, X_ctx, industry_code, v_prev, mask):
            B, N = mask.shape
        
            industry_embed = self.industry_embedding(industry_code)
            was_bottom = X_ctx[:, :, 0]
        
            gamma_clamped = torch.clamp(self.gamma, min=0.0)
            delta_clamped = torch.clamp(self.delta, min=0.0)
            momentum_term = gamma_clamped * v_prev.unsqueeze(-1)
        
            mlp_input = torch.cat([X_pop, industry_embed, X_ctx, momentum_term], dim=-1)
        
            h = self.prior_mlp(mlp_input)
            mu_base = self.mu_head(h).squeeze(-1)
            logvar = self.logvar_head(h).squeeze(-1)
        
            mu = mu_base + delta_clamped * was_bottom
        
            if self.training:
                std = torch.exp(0.5 * logvar)
                eps = torch.randn_like(std)
                z = mu + eps * std
            else:
                z = mu
        
            z_masked = z.masked_fill(mask == 0, -1e9)
            v_hat = F.softmax(z_masked, dim=-1)
        
            return v_hat, mu, logvar

    class SeasonAwareFusion(nn.Module):
        """èµ›å­£æ„ŸçŸ¥èåˆå±‚"""
        def __init__(self, num_seasons: int = 34):
            super().__init__()
            rank_seasons = set([1, 2] + list(range(28, num_seasons + 1)))
            alpha_init = torch.full((num_seasons,), 0.5)
            g_init = torch.full((num_seasons,), 0.5)
            for season in range(1, num_seasons + 1):
                if season in rank_seasons:
                    alpha_init[season - 1] = 0.55
                    g_init[season - 1] = 0.2
                else:
                    alpha_init[season - 1] = 0.5
                    g_init[season - 1] = 0.8
            self.alpha_s = nn.Parameter(torch.logit(alpha_init.clamp(0.01, 0.99)))
            self.g_s = nn.Parameter(torch.logit(g_init.clamp(0.01, 0.99)))
    
        def forward(self, X_skill, v_hat, season_ids, mask, is_test=False):
            B, N = mask.shape
        
            pct_J = X_skill[:, :, 0]
            rank_J = X_skill[:, :, 2]
        
            N_t = mask.sum(dim=1, keepdim=True)
            R_J = (N_t - rank_J) / (N_t - 1 + 1e-8)
            R_J = R_J * mask
        
            v_hat_masked = v_hat.masked_fill(mask == 0, -1e9)
            rank_V = torch.argsort(torch.argsort(v_hat_masked, dim=1, descending=True), dim=1).float()
            rank_V = rank_V + 1
            R_V = (N_t - rank_V) / (N_t - 1 + 1e-8)
            R_V = R_V * mask
        
            if is_test:
                alpha_val = torch.sigmoid(self.alpha_s).mean()
                g_val = torch.sigmoid(self.g_s).mean()
                alpha = alpha_val.expand(B)
                g = g_val.expand(B)
            else:
                alpha = torch.sigmoid(self.alpha_s[season_ids])
                g = torch.sigmoid(self.g_s[season_ids])
        
            S_pct = alpha.unsqueeze(1) * pct_J + (1 - alpha.unsqueeze(1)) * v_hat
            S_rank = alpha.unsqueeze(1) * R_J + (1 - alpha.unsqueeze(1)) * R_V
            S_final = g.unsqueeze(1) * S_pct + (1 - g.unsqueeze(1)) * S_rank
            S_final = S_final * mask
        
            return S_final

    class DWTSModel(nn.Module):
        """å®Œæ•´æ¨¡å‹"""
        def __init__(self, skill_dim, pop_dim, ctx_dim, hidden_dim, num_seasons, tau,
                     num_industries, industry_embed_dim):
            super().__init__()
        
            self.latent_engine = LatentInferenceEngine(
                pop_dim, ctx_dim, hidden_dim, num_industries, industry_embed_dim
            )
            self.fusion_layer = SeasonAwareFusion(num_seasons)
            self.tau = nn.Parameter(torch.tensor(float(tau)), requires_grad=True)
    
        def forward(self, batch, v_prev=None, is_test=False):
            X_skill = batch['X_skill']
            X_pop = batch['X_pop']
            X_ctx = batch['X_ctx']
            industry_code = batch['industry_code']
            mask = batch['mask']
            season_ids = batch['season'] - 1
            week = batch['week'][0].item()
        
            if v_prev is None or week == 1:
                N_t = mask.sum(dim=1, keepdim=True)
                v_prev = mask / N_t
        
            v_hat, mu, logvar = self.latent_engine(X_pop, X_ctx, industry_code, v_prev, mask)
            S_final = self.fusion_layer(X_skill, v_hat, season_ids, mask, is_test=is_test)
        
            S_masked = S_final.masked_fill(mask == 0, 1e9)
            tau_clamped = torch.clamp(self.tau, min=0.05, max=0.5)
            p_elim = F.softmax(-S_masked / tau_clamped, dim=-1)
        
            p_bottom = torch.sigmoid(-S_final)
            p_bottom = p_bottom * mask
        
            return {
                'p_elim': p_elim,
                'p_bottom': p_bottom,
                'v_hat': v_hat,
                'mu': mu,
                'logvar': logvar,
                'S_final': S_final,
            }

    # ==================== æŸå¤±å‡½æ•° ====================
    class CompositeLoss(nn.Module):
        """å¤åˆæŸå¤±å‡½æ•°"""
        def __init__(self, lambda_bottom, lambda_top2, lambda_kl, lambda_smooth):
            super().__init__()
            self.lambda_bottom = lambda_bottom
            self.lambda_top2 = lambda_top2
            self.lambda_kl = lambda_kl
            self.lambda_smooth = lambda_smooth
    
        def forward(self, outputs, batch, v_hat_prev=None):
            p_elim = outputs['p_elim']
            p_bottom = outputs['p_bottom']
            v_hat = outputs['v_hat']
            mu = outputs['mu']
            logvar = outputs['logvar']
        
            y_elim = batch['y_elim']
            y_bottom = batch['y_bottom']
            mask = batch['mask']
        
            elim_mask = (y_elim == 1) & (mask == 1)
            if elim_mask.sum() > 0:
                L_elim = -torch.log(p_elim[elim_mask] + 1e-10).mean()
            else:
                L_elim = torch.tensor(0.0, device=p_elim.device)
        
            active_mask = mask == 1
            if active_mask.sum() > 0:
                L_bottom = F.binary_cross_entropy(
                    p_bottom[active_mask],
                    y_bottom[active_mask],
                    reduction='mean'
                )
            else:
                L_bottom = torch.tensor(0.0, device=p_bottom.device)
        
            if active_mask.sum() > 0:
                kl_div = -0.5 * (1 + logvar - mu.pow(2) - logvar.exp())
                kl_per_sample = kl_div[active_mask].sum()
                L_KL = kl_per_sample / active_mask.sum()
            else:
                L_KL = torch.tensor(0.0, device=mu.device)
        
            if active_mask.sum() > 0:
                L_top2 = torch.tensor(0.0, device=p_elim.device)
                for i in range(p_elim.size(0)):
                    bottom_mask = (y_bottom[i] == 1) & (mask[i] == 1)
                    if bottom_mask.sum() == 0:
                        continue
                    bottom_mass = p_elim[i][bottom_mask].sum()
                    L_top2 = L_top2 - torch.log(bottom_mass + 1e-10)
                L_top2 = L_top2 / p_elim.size(0)
            else:
                L_top2 = torch.tensor(0.0, device=p_elim.device)

            if v_hat_prev is not None and active_mask.sum() > 0:
                smooth_diff = (v_hat - v_hat_prev).pow(2)
                L_smooth = smooth_diff[active_mask].mean()
            else:
                L_smooth = torch.tensor(0.0, device=v_hat.device)
        
            total_loss = (
                L_elim +
                self.lambda_bottom * L_bottom +
                self.lambda_top2 * L_top2 +
                self.lambda_kl * L_KL +
                self.lambda_smooth * L_smooth
            )
        
            loss_dict = {
                'total': total_loss.item(),
                'elim': L_elim.item(),
                'bottom': L_bottom.item(),
                'kl': L_KL.item(),
                'smooth': L_smooth.item(),
            'top2': L_top2.item(),
            }
        
            return total_loss, loss_dict

    # ==================== ğŸš© æŒ‡ä»¤ 1ï¼šå‡†å¤‡å…¨é‡æ•°æ®ä¸æ¨¡å‹ ====================
    print("\n" + "=" * 80)
    print("ğŸš© æŒ‡ä»¤ 1ï¼šå‡†å¤‡å…¨é‡æ•°æ®ä¸æ¨¡å‹")
    print("=" * 80)

    # åŠ è½½åŸå§‹æ•°æ®
    df_raw = pd.read_csv(DATA_PATH)
    print(f"âœ“ åŠ è½½åŸå§‹æ•°æ®: {len(df_raw)} è¡Œ")

    # æ•°å€¼ç‰¹å¾æ ‡å‡†åŒ–
    numerical_features = [
        'relative_score_pct', 'score_deviation', 'weekly_rank',
        'search_mean_log', 'search_recent_3m_log', 'total_followers_log',
        'weeks_survived', 'running_order'
    ]

    scaler = StandardScaler()
    df_raw[numerical_features] = scaler.fit_transform(df_raw[numerical_features])
    print(f"âœ“ ç‰¹å¾æ ‡å‡†åŒ–å®Œæˆ")

    # åˆ›å»ºæ•°æ®é›†
    full_dataset = DWTSDataset(df_raw, N_max=HYPERPARAMS['N_max'], verbose=True)

    # åˆ›å»ºä¸¤ä¸ªåŠ è½½å™¨
    train_loader = DataLoader(full_dataset, batch_size=1, shuffle=True)
    extract_loader = DataLoader(full_dataset, batch_size=1, shuffle=False)

    print(f"âœ“ æ•°æ®åŠ è½½å™¨åˆ›å»ºå®Œæˆ:")
    print(f"  - train_loader: shuffle=True (ç”¨äºè®­ç»ƒ)")
    print(f"  - extract_loader: shuffle=False (ç”¨äºæå–)")

    # é‡æ–°å®ä¾‹åŒ–æ¨¡å‹
    model = DWTSModel(
        skill_dim=3,
        pop_dim=3,
        ctx_dim=3,
        hidden_dim=HYPERPARAMS['hidden_dim'],
        num_seasons=34,
        tau=HYPERPARAMS['tau'],
        num_industries=HYPERPARAMS['num_industries'],
        industry_embed_dim=HYPERPARAMS['industry_embed_dim']
    ).to(DEVICE)

    print(f"âœ“ æ¨¡å‹å®ä¾‹åŒ–å®Œæˆ")
    print(f"  - å‚æ•°é‡: {sum(p.numel() for p in model.parameters()):,}")

    # ==================== ğŸš© æŒ‡ä»¤ 2ï¼šæ‰§è¡Œå…¨é‡è®­ç»ƒ ====================
    print("\n" + "=" * 80)
    print("ğŸš© æŒ‡ä»¤ 2ï¼šæ‰§è¡Œå…¨é‡è®­ç»ƒ (Full Training)")
    print("=" * 80)

    # åˆå§‹åŒ–ä¼˜åŒ–å™¨å’ŒæŸå¤±å‡½æ•°
    optimizer = torch.optim.Adam(model.parameters(), lr=HYPERPARAMS['learning_rate'])
    criterion = CompositeLoss(
        lambda_bottom=HYPERPARAMS['lambda_bottom'],
        lambda_top2=HYPERPARAMS['lambda_top2'],
        lambda_kl=HYPERPARAMS['lambda_kl'],
        lambda_smooth=HYPERPARAMS['lambda_smooth']
    )

    print(f"å¼€å§‹è®­ç»ƒ {HYPERPARAMS['num_epochs']} ä¸ª Epoch...")

    for epoch in range(HYPERPARAMS['num_epochs']):
        model.train()
        total_loss = 0
        loss_components = {'elim': 0, 'bottom': 0, 'top2': 0, 'kl': 0, 'smooth': 0}
        num_batches = 0
    
        v_hat_cache = {}
    
        for batch in train_loader:
            batch = {k: v.to(DEVICE) if isinstance(v, torch.Tensor) else v 
                    for k, v in batch.items()}
        
            season = batch['season'][0].item()
            week = batch['week'][0].item()
        
            key = (season, week - 1)
            v_prev = v_hat_cache.get(key, None)
        
            outputs = model(batch, v_prev=v_prev, is_test=False)
            loss, loss_dict = criterion(outputs, batch, v_hat_prev=v_prev)
        
            optimizer.zero_grad()
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer.step()
        
            v_hat_cache[(season, week)] = outputs['v_hat'].detach()
        
            total_loss += loss.item()
            for k in loss_components:
                loss_components[k] += loss_dict.get(k, 0)
            num_batches += 1
    
        avg_loss = total_loss / num_batches
        avg_components = {k: v / num_batches for k, v in loss_components.items()}
    
        if (epoch + 1) % 10 == 0:
            print(f"Epoch {epoch+1}/{HYPERPARAMS['num_epochs']}: "
                  f"Loss={avg_loss:.4f}, "
                  f"elim={avg_components['elim']:.4f}, "
                  f"bottom={avg_components['bottom']:.4f}, "
                  f"top2={avg_components['top2']:.4f}, "
                  f"kl={avg_components['kl']:.4f}, "
                  f"smooth={avg_components['smooth']:.4f}")

    print(f"âœ“ å…¨é‡è®­ç»ƒå®Œæˆ")

    # ==================== ğŸš© æŒ‡ä»¤ 3ï¼šæå–éšå˜é‡ (v_hat) ====================
    print("\n" + "=" * 80)
    print("ğŸš© æŒ‡ä»¤ 3ï¼šæå–éšå˜é‡ (v_hat)")
    print("=" * 80)

    model.eval()

    # ğŸ”§ å…³é”®ä¿®å¤ï¼šä½¿ç”¨ä¸¤ä¸ªç¼“å­˜
    v_hat_tensor_cache = {}  # ç”¨äºè·¨å‘¨ä¼ é€’ï¼ˆTensor æ ¼å¼ï¼‰
    v_hat_numpy_cache = {}   # ç”¨äºæœ€ç»ˆå¯¼å‡ºï¼ˆnumpy æ ¼å¼ï¼‰

    print(f"å¼€å§‹é¡ºåºæ¨æ–­...")

    with torch.no_grad():
        for batch_idx, batch in enumerate(extract_loader):
            batch = {k: v.to(DEVICE) if isinstance(v, torch.Tensor) else v 
                    for k, v in batch.items()}
        
            season = batch['season'][0].item()
            week = batch['week'][0].item()
            N_actual = batch['N_actual'][0].item()
        
            key_prev = (season, week - 1)
            # ğŸ”§ ä» Tensor ç¼“å­˜ä¸­è·å–ä¸Šä¸€å‘¨çš„ v_hat
            v_prev = v_hat_tensor_cache.get(key_prev, None)
        
            # å¼€å¯æµ‹è¯•æ¨¡å¼
            outputs = model(batch, v_prev=v_prev, is_test=True)
        
            # ğŸ”§ åŒæ—¶å­˜å‚¨ä¸¤ç§æ ¼å¼
            key_current = (season, week)
            v_hat_tensor_cache[key_current] = outputs['v_hat'].detach()  # Tensor æ ¼å¼ï¼ˆç”¨äºä¸‹ä¸€å‘¨ï¼‰
            v_hat_numpy_cache[key_current] = outputs['v_hat'].cpu().numpy()[0]  # numpy æ ¼å¼ï¼ˆç”¨äºå¯¼å‡ºï¼‰
        
            if (batch_idx + 1) % 100 == 0:
                print(f"  å·²å¤„ç† {batch_idx + 1}/{len(extract_loader)} ä¸ªæ‰¹æ¬¡")

    print(f"âœ“ éšå˜é‡æå–å®Œæˆ: {len(v_hat_numpy_cache)} ä¸ª (season, week) ç»„")

    # ==================== ğŸš© æŒ‡ä»¤ 4ï¼šæ•°æ®å¯¹é½ä¸å¯¼å‡º ====================
    print("\n" + "=" * 80)
    print("ğŸš© æŒ‡ä»¤ 4ï¼šæ•°æ®å¯¹é½ä¸å¯¼å‡º")
    print("=" * 80)

    # é‡æ–°åŠ è½½åŸå§‹æ•°æ®ï¼ˆæœªæ ‡å‡†åŒ–çš„ç‰ˆæœ¬ï¼‰
    df_output = pd.read_csv(DATA_PATH)
    df_output = df_output.sort_values(['season', 'week', 'celebrity_name']).reset_index(drop=True)

    print(f"åŸå§‹æ•°æ®è¡Œæ•°: {len(df_output)}")

    # åˆ›å»ºæ–°åˆ—
    df_output['estimated_fan_vote'] = np.nan

    # æŒ‰ (season, week) åˆ†ç»„å›å¡«
    matched_count = 0
    unmatched_count = 0

    for (season, week), group_df in df_output.groupby(['season', 'week']):
        key = (season, week)
    
        # ğŸ”§ ä» numpy ç¼“å­˜ä¸­è·å–
        if key in v_hat_numpy_cache:
            v_hat_full = v_hat_numpy_cache[key]  # shape: (20,)
            N_actual = len(group_df)
        
            # æˆªæ–­åˆ°å®é™…é€‰æ‰‹æ•°
            v_hat_truncated = v_hat_full[:N_actual]
        
            # å›å¡«åˆ° DataFrame
            df_output.loc[group_df.index, 'estimated_fan_vote'] = v_hat_truncated
            matched_count += len(group_df)
        else:
            unmatched_count += len(group_df)
            print(f"è­¦å‘Š: æœªæ‰¾åˆ° Season {season} Week {week} çš„ v_hat")

    print(f"\nâœ“ æ•°æ®å¯¹é½å®Œæˆ:")
    print(f"  - æˆåŠŸåŒ¹é…: {matched_count} è¡Œ")
    print(f"  - æœªåŒ¹é…: {unmatched_count} è¡Œ")

    # éªŒè¯
    missing_count = df_output['estimated_fan_vote'].isna().sum()
    print(f"  - ç¼ºå¤±å€¼: {missing_count} è¡Œ")

    # ä¿å­˜æ–‡ä»¶
    df_output.to_csv(OUTPUT_PATH, index=False)
    print(f"\nâœ“ æ–‡ä»¶å·²ä¿å­˜: {OUTPUT_PATH}")

    # ==================== æœ€ç»ˆéªŒè¯ ====================
    print("\n" + "=" * 80)
    print("âœ… æœ€ç»ˆéªŒè¯")
    print("=" * 80)

    print(f"\nğŸ“Š æ•°æ®ç»Ÿè®¡:")
    print(f"  - æ€»è¡Œæ•°: {len(df_output)}")
    print(f"  - æ–°åˆ—å: estimated_fan_vote")
    print(f"  - éç©ºå€¼: {df_output['estimated_fan_vote'].notna().sum()}")
    print(f"  - å€¼åŸŸ: [{df_output['estimated_fan_vote'].min():.6f}, {df_output['estimated_fan_vote'].max():.6f}]")
    print(f"  - å‡å€¼: {df_output['estimated_fan_vote'].mean():.6f}")

    # å±•ç¤ºæ ·ä¾‹
    print(f"\nğŸ“‹ æ•°æ®æ ·ä¾‹ (Season 30, Week 1):")
    sample = df_output[(df_output['season'] == 30) & (df_output['week'] == 1)][
        ['celebrity_name', 'weekly_total', 'estimated_fan_vote']
    ].head(10)
    print(sample.to_string(index=False))

    # éªŒè¯æ¦‚ç‡å’Œ
    print(f"\nğŸ” æ¦‚ç‡å’ŒéªŒè¯ (æ¯å‘¨åº”æ¥è¿‘ 1.0):")
    prob_sums = df_output.groupby(['season', 'week'])['estimated_fan_vote'].sum()
    print(f"  - æœ€å°å€¼: {prob_sums.min():.6f}")
    print(f"  - æœ€å¤§å€¼: {prob_sums.max():.6f}")
    print(f"  - å‡å€¼: {prob_sums.mean():.6f}")

    if abs(prob_sums.mean() - 1.0) < 0.01:
        print(f"  âœ… æ¦‚ç‡å’ŒéªŒè¯é€šè¿‡ï¼")
    else:
        print(f"  âš ï¸  æ¦‚ç‡å’Œåç¦» 1.0ï¼Œè¯·æ£€æŸ¥")

    print("\n" + "=" * 80)




    # ==================== ğŸš© æŒ‡ä»¤ 4.5ï¼šä¸ç¡®å®šæ€§åº¦é‡ ====================
    print("\n" + "=" * 80)
    print("ğŸš© æŒ‡ä»¤ 4.5ï¼šä¸ç¡®å®šæ€§åº¦é‡")
    print("=" * 80)

    # ä½¿ç”¨ç»„å†…æ³¢åŠ¨ä¼°è®¡ä¸ç¡®å®šæ€§åŒºé—´å®½åº¦
    uncertainty_df = df_output.copy()
    uncertainty_df['vote_std'] = uncertainty_df.groupby(['season', 'week'])['estimated_fan_vote'].transform('std')
    uncertainty_df['uncertainty_width'] = 1.96 * uncertainty_df['vote_std'].fillna(0)

    print(f"\nğŸ“ ä¸ç¡®å®šæ€§åŒºé—´å®½åº¦å‡å€¼: {uncertainty_df['uncertainty_width'].mean():.4f}")

    # å‘¨æ¬¡ä¸ç¡®å®šæ€§åˆ†å¸ƒ (Top 10)
    weekly_uncertainty = (
        uncertainty_df.groupby(['season', 'week'])['uncertainty_width']
        .mean()
        .sort_values(ascending=False)
        .head(10)
    )
    print("\nâš ï¸ å‘¨æ¬¡ä¸ç¡®å®šæ€§ Top 10:")
    print(weekly_uncertainty)

    # é€‰æ‰‹ä¸ç¡®å®šæ€§åˆ†å¸ƒ (Top 10)
    player_uncertainty = (
        uncertainty_df.groupby('celebrity_name')['uncertainty_width']
        .mean()
        .sort_values(ascending=False)
        .head(10)
    )
    print("\nâš ï¸ é€‰æ‰‹ä¸ç¡®å®šæ€§ Top 10:")
    print(player_uncertainty)
    # # ==================== ğŸš© æŒ‡ä»¤ 5ï¼šæ ¡å‡†ä¸åŸºçº¿å¯¹æ¯” ====================
    # print("\n" + "=" * 80)
    # print("ğŸš© æŒ‡ä»¤ 5ï¼šæ ¡å‡†ä¸åŸºçº¿å¯¹æ¯”")
    # print("=" * 80)

    # # 1. æ„å»ºåŸºçº¿ï¼šæŒ‰è¯„å§”åˆ†å æ¯”ä½œä¸ºç²‰ä¸ç¥¨æ›¿ä»£
    # baseline_df = df_output.copy()
    # baseline_df['baseline_fan_vote'] = baseline_df.groupby(['season', 'week'])['weekly_total'].transform(lambda x: x / x.sum())

    # # è¿½åŠ åŸºçº¿: æ’ååˆ¶åˆ†æ•° & æŒ‡æ•°åŠ æƒç²‰ä¸ç¥¨
    # baseline_df['rank_score'] = baseline_df.groupby(['season', 'week'])['weekly_total'].rank(ascending=False, method='min')
    # baseline_df['rank_score'] = baseline_df.groupby(['season', 'week'])['rank_score'].transform(
    #     lambda x: (x.max() - x) / (x.max() - 1 + 1e-8)
    # )

    # baseline_df['blend_score'] = 0.5 * baseline_df['baseline_fan_vote'] + 0.5 * baseline_df['rank_score']
    # baseline_df['baseline_log_vote'] = np.log1p(baseline_df['baseline_fan_vote'])


    # # 2. æ¦‚ç‡å’Œæ ¡éªŒï¼ˆæ¨¡å‹ vs åŸºçº¿ï¼‰
    # model_sums = df_output.groupby(['season', 'week'])['estimated_fan_vote'].sum()
    # baseline_sums = baseline_df.groupby(['season', 'week'])['baseline_fan_vote'].sum()

    # print(f"\nğŸ” æ¦‚ç‡å’Œå¯¹æ¯”:")
    # print(f"  - æ¨¡å‹ç¥¨å’Œ: æœ€å° {model_sums.min():.4f} | æœ€å¤§ {model_sums.max():.4f} | å‡å€¼ {model_sums.mean():.4f}")
    # print(f"  - åŸºçº¿ç¥¨å’Œ: æœ€å° {baseline_sums.min():.4f} | æœ€å¤§ {baseline_sums.max():.4f} | å‡å€¼ {baseline_sums.mean():.4f}")

    # # 3. å¤šæŒ‡æ ‡è¯„ä¼°ï¼šæ¨¡å‹ vs åŸºçº¿

    # def _rank_metrics(group, score_col, k=2):
    #     if group['is_eliminated'].sum() == 0:
    #         return None
    #     true_elim = set(group[group['is_eliminated'] == 1]['celebrity_name'].values)
    #     ranked = group.sort_values(score_col).reset_index(drop=True)
    #     topk = set(ranked.head(k)['celebrity_name'].values)
    #     top1 = set(ranked.head(1)['celebrity_name'].values)
    #     hit_topk = 1 if (true_elim & topk) else 0
    #     hit_top1 = 1 if (true_elim & top1) else 0
    #     if len(true_elim) > 0:
    #         true_rank = ranked.index[ranked['celebrity_name'].isin(true_elim)].min() + 1
    #     else:
    #         true_rank = None
    #     mean_elim_score = group.loc[group['is_eliminated'] == 1, score_col].mean()
    #     mean_safe_score = group.loc[group['is_eliminated'] == 0, score_col].mean()
    #     return {
    #         'hit_top1': hit_top1,
    #         'hit_topk': hit_topk,
    #         'true_rank': true_rank,
    #         'mean_elim_score': mean_elim_score,
    #         'mean_safe_score': mean_safe_score,
    #     }

    # metrics = {
    #     'model': {
    #         'hit_top1': 0,
    #         'hit_top2': 0,
    #         'rank_sum': 0,
    #         'count': 0,
    #         'score_gap_sum': 0.0,
    #     },
    #     'baseline_vote': {
    #         'hit_top1': 0,
    #         'hit_top2': 0,
    #         'rank_sum': 0,
    #         'count': 0,
    #         'score_gap_sum': 0.0,
    #     },
    #     'baseline_rank': {
    #         'hit_top1': 0,
    #         'hit_top2': 0,
    #         'rank_sum': 0,
    #         'count': 0,
    #         'score_gap_sum': 0.0,
    #     },
    #     'baseline_blend': {
    #         'hit_top1': 0,
    #         'hit_top2': 0,
    #         'rank_sum': 0,
    #         'count': 0,
    #         'score_gap_sum': 0.0,
    #     },
    #     'baseline_log': {
    #         'hit_top1': 0,
    #         'hit_top2': 0,
    #         'rank_sum': 0,
    #         'count': 0,
    #         'score_gap_sum': 0.0,
    #     }
    # }

    # for (season, week), group in df_output.groupby(['season', 'week']):
    #     model_result = _rank_metrics(group, 'estimated_fan_vote', k=2)
    #     baseline_variants = {
    #         'baseline_vote': 'baseline_fan_vote',
    #         'baseline_rank': 'rank_score',
    #         'baseline_blend': 'blend_score',
    #         'baseline_log': 'baseline_log_vote',
    #     }
    #     baseline_results = {
    #         name: _rank_metrics(baseline_df.loc[group.index], col, k=2)
    #         for name, col in baseline_variants.items()
    #     }

    #     if model_result is None or any(result is None for result in baseline_results.values()):
    #         continue
    #     metrics['model']['hit_top1'] += model_result['hit_top1']
    #     metrics['model']['hit_top2'] += model_result['hit_topk']
    #     metrics['model']['rank_sum'] += model_result['true_rank']
    #     metrics['model']['count'] += 1
    #     metrics['model']['score_gap_sum'] += model_result['mean_safe_score'] - model_result['mean_elim_score']
    
    #     for baseline_name, baseline_result in baseline_results.items():
    #         metrics[baseline_name]['hit_top1'] += baseline_result['hit_top1']
    #         metrics[baseline_name]['hit_top2'] += baseline_result['hit_topk']
    #         metrics[baseline_name]['rank_sum'] += baseline_result['true_rank']
    #         metrics[baseline_name]['count'] += 1
    #         metrics[baseline_name]['score_gap_sum'] += baseline_result['mean_safe_score'] - baseline_result['mean_elim_score']

    # print("\nğŸ¯ å¤šæŒ‡æ ‡è¯„ä¼°ç»“æœ (æ¨¡å‹ vs å¤šåŸºçº¿):")
    # for name, stat in metrics.items():
    #     count = stat['count'] or 1
    #     print(f"\n{name.upper()}:")
    #     print(f"  - Top-1 å‘½ä¸­ç‡: {stat['hit_top1']/count:.2%} ({stat['hit_top1']}/{count})")
    #     print(f"  - Top-2 å‘½ä¸­ç‡: {stat['hit_top2']/count:.2%} ({stat['hit_top2']}/{count})")
    #     print(f"  - çœŸå®æ·˜æ±°å¹³å‡åæ¬¡: {stat['rank_sum']/count:.2f}")
    #     print(f"  - æ·˜æ±°/å®‰å…¨åˆ†å·®å‡å€¼: {stat['score_gap_sum']/count:.4f}")


    # # 3.5 ç²‰ä¸ç¥¨æ¨æ–­è´¨é‡è¯„ä¼° (Kendall Ï„ / MAE / MSE / äº‰è®®å‘¨æ¬¡)
    # def _kendall_tau(x, y):
    #     x = np.asarray(x)
    #     y = np.asarray(y)
    #     n = len(x)
    #     if n < 2:
    #         return np.nan
    #     concordant = 0
    #     discordant = 0
    #     ties_x = 0
    #     ties_y = 0
    #     for i in range(n - 1):
    #         for j in range(i + 1, n):
    #             dx = x[i] - x[j]
    #             dy = y[i] - y[j]
    #             if dx == 0 and dy == 0:
    #                 continue
    #             if dx == 0:
    #                 ties_x += 1
    #                 continue
    #             if dy == 0:
    #                 ties_y += 1
    #                 continue
    #             if dx * dy > 0:
    #                 concordant += 1
    #             elif dx * dy < 0:
    #                 discordant += 1
    #     denom = ((concordant + discordant + ties_x) * (concordant + discordant + ties_y)) ** 0.5
    #     if denom == 0:
    #         return np.nan
    #     return (concordant - discordant) / denom

    # kendall_scores = []
    # mae_scores = []
    # mse_scores = []

    # for (season, week), group in df_output.groupby(['season', 'week']):
    #     baseline_group = baseline_df.loc[group.index]
    #     if group['estimated_fan_vote'].isna().any() or baseline_group['baseline_fan_vote'].isna().any():
    #         continue
    #     tau = _kendall_tau(group['estimated_fan_vote'], baseline_group['baseline_fan_vote'])
    #     if tau == tau:  # è¿‡æ»¤ NaN
    #         kendall_scores.append(tau)
    #     diff = group['estimated_fan_vote'] - baseline_group['baseline_fan_vote']
    #     mae_scores.append(diff.abs().mean())
    #     mse_scores.append((diff ** 2).mean())

    # print("\nğŸ“ ç²‰ä¸ç¥¨æ¨æ–­è´¨é‡ (ä¸è¯„å§”å æ¯”åŸºçº¿å¯¹æ¯”):")
    # print(f"  - Kendall Ï„ å‡å€¼: {np.mean(kendall_scores):.4f}")
    # print(f"  - MAE å‡å€¼: {np.mean(mae_scores):.4f}")
    # print(f"  - MSE å‡å€¼: {np.mean(mse_scores):.4f}")

    # # äº‰è®®å‘¨æ¬¡åå·®åˆ†æï¼ˆä½¿ç”¨å·²çŸ¥äº‰è®®èµ›å­£ï¼‰
    # controversy_cases = [
    #     {'season': 2, 'celebrity': 'Jerry', 'label': 'Season 2 Jerry Rice'},
    #     {'season': 4, 'celebrity': 'Billy', 'label': 'Season 4 Billy Ray Cyrus'},
    #     {'season': 11, 'celebrity': 'Bristol', 'label': 'Season 11 Bristol Palin'},
    #     {'season': 27, 'celebrity': 'Bobby', 'label': 'Season 27 Bobby Bones'},
    # ]

    # controversy_rows = []
    # for case in controversy_cases:
    #     season = case['season']
    #     name = case['celebrity']
    #     subset = df_output[(df_output['season'] == season) & (df_output['celebrity_name'].str.contains(name, case=False, na=False))]
    #     if subset.empty:
    #         continue
    #     for week in subset['week'].unique():
    #         week_group = df_output[(df_output['season'] == season) & (df_output['week'] == week)].copy()
    #         base_group = baseline_df.loc[week_group.index]
    #         week_group['model_rank'] = week_group['estimated_fan_vote'].rank(ascending=False, method='min')
    #         week_group['baseline_rank'] = base_group['baseline_fan_vote'].rank(ascending=False, method='min')
    #         week_group['rank_gap'] = (week_group['model_rank'] - week_group['baseline_rank']).abs()
    #         focus = week_group[week_group['celebrity_name'].str.contains(name, case=False, na=False)].copy()
    #         focus['case'] = case['label']
    #         controversy_rows.append(focus)

    # if controversy_rows:
    #     controversy_df = pd.concat(controversy_rows).sort_values('rank_gap', ascending=False)
    #     print("\nâš ï¸ äº‰è®®å‘¨æ¬¡åå·® Top 10 (æ¨¡å‹æ’å vs åŸºçº¿æ’å):")
    #     print(controversy_df[['case', 'season', 'week', 'celebrity_name', 'model_rank', 'baseline_rank', 'rank_gap']].head(10))
    # else:
    #     print("\nâš ï¸ æœªæ‰¾åˆ°äº‰è®®å‘¨æ¬¡æ ·æœ¬")
    # # 4. è¯„å§”åˆ† vs ç²‰ä¸ç¥¨å·®å¼‚(äº‰è®®ä¿¡å·)ç»Ÿè®¡
    # vote_gap = df_output.copy()
    # vote_gap['judge_share'] = vote_gap.groupby(['season', 'week'])['weekly_total'].transform(lambda x: x / x.sum())
    # vote_gap['vote_gap'] = (vote_gap['estimated_fan_vote'] - vote_gap['judge_share']).abs()

    # season_gap = vote_gap.groupby('season')['vote_gap'].mean().sort_values(ascending=False).head(5)
    # print(f"\nğŸ”¥ è¯„å§”/ç²‰ä¸å·®å¼‚æœ€å¤§çš„èµ›å­£ (Top 5):")
    # print(season_gap)
    # print("ğŸ‰ æ‰€æœ‰æŒ‡ä»¤æ‰§è¡Œå®Œæˆï¼")
    # print("=" * 80)
# Task 1 (new framework)
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# ==================== é…ç½®åŒº ====================
DATA_PATH = './panel_data_clean_v2.csv'
OUTPUT_PATH = './estimated_fan_votes.csv'
TOTAL_VOTES = 1_000_000
RANKING_SEASONS = {1, 2, 28, 29, 30, 31, 32, 33, 34}

print("=" * 80)
print("ğŸ“Š Task 1: è§‚ä¼—æŠ•ç¥¨åæ¨æ¨¡å‹ (çº¦æŸä¼˜åŒ–æ¡†æ¶)")
print("=" * 80)

# æ•°æ®åŠ è½½
panel_df = pd.read_csv(DATA_PATH)
print(f"âœ“ æ•°æ®åŠ è½½å®Œæˆ: {len(panel_df)} è¡Œ")


def _prepare_popularity(group: pd.DataFrame) -> pd.Series:
    pop_base = group.get('total_followers_log', pd.Series(0, index=group.index)).fillna(0)
    pop_base = pop_base + group.get('search_mean_log', pd.Series(0, index=group.index)).fillna(0)
    pop_base = pop_base.replace(0, np.nan)
    pop_base = pop_base.fillna(pop_base.median() if pd.notna(pop_base.median()) else 1)
    pop_base = pop_base - pop_base.min() + 1
    return pop_base


def infer_rank_group(group: pd.DataFrame) -> pd.DataFrame:
    group = group.copy()
    n_contestants = len(group)
    group['judge_rank'] = group['weekly_total'].rank(ascending=False, method='min')

    elim_idx = group.index[group['is_eliminated'] == 1].tolist()
    elim_idx = elim_idx[0] if elim_idx else None

    vote_rank = pd.Series(index=group.index, dtype=float)
    others = group.index.difference([elim_idx]) if elim_idx is not None else group.index
    others_sorted = group.loc[others].sort_values('judge_rank', ascending=False).index

    for rank_value, idx in enumerate(others_sorted, start=1):
        vote_rank[idx] = rank_value
    if elim_idx is not None:
        vote_rank[elim_idx] = n_contestants

    group['vote_rank'] = vote_rank
    group['total_rank'] = group['judge_rank'] + group['vote_rank']

    vote_score = n_contestants - group['vote_rank'] + 1
    group['vote_share'] = vote_score / vote_score.sum()
    group['estimated_fan_vote'] = group['vote_share']
    group['estimated_vote_count'] = group['vote_share'] * TOTAL_VOTES
    return group


def infer_percent_group(group: pd.DataFrame) -> pd.DataFrame:
    group = group.copy()
    group['judge_pct'] = group['weekly_total'] / group['weekly_total'].sum()

    pop_base = _prepare_popularity(group)
    base_share = pop_base / pop_base.sum()

    elim_idx = group.index[group['is_eliminated'] == 1].tolist()
    elim_idx = elim_idx[0] if elim_idx else None

    if elim_idx is not None:
        elim_judge = group.loc[elim_idx, 'judge_pct']
        margin = 1e-6
        required = (elim_judge - group['judge_pct'] + margin).clip(lower=0)
        required.loc[elim_idx] = 0
        required_sum = required.sum()

        if required_sum < 1:
            remaining = 1 - required_sum
            weights = base_share.copy()
            weights.loc[elim_idx] = 0
            if weights.sum() == 0:
                weights = pd.Series(1, index=group.index)
                weights.loc[elim_idx] = 0
            weights = weights / weights.sum()
            vote_share = required + remaining * weights
        else:
            vote_share = required / required_sum

        vote_share.loc[elim_idx] = max(0, 1 - vote_share.drop(elim_idx).sum())
    else:
        vote_share = base_share

    group['vote_share'] = vote_share
    group['estimated_fan_vote'] = group['vote_share']
    group['estimated_vote_count'] = group['vote_share'] * TOTAL_VOTES
    group['total_pct'] = group['judge_pct'] + group['vote_share']
    return group


results = []
for (season, week), group in panel_df.groupby(['season', 'week']):
    if season in RANKING_SEASONS:
        results.append(infer_rank_group(group))
    else:
        results.append(infer_percent_group(group))

result_df = pd.concat(results, ignore_index=True)
result_df = result_df.sort_values(['season', 'week', 'celebrity_name']).reset_index(drop=True)

result_df.to_csv(OUTPUT_PATH, index=False)
print(f"âœ… Task 1 åæ¨ç»“æœå·²ä¿å­˜: {OUTPUT_PATH}")
```

```python
#Task 1éªŒè¯ + Task 2
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# ==================== é…ç½®åŒº ====================
DATA_PATH = './estimated_fan_votes.csv'
OUTPUT_DIR = './'

print("=" * 80)
print("ğŸ¯ DWTS æœºåˆ¶é‡æ„ä¸åäº‹å®æ¨¡æ‹Ÿ (ä¿®å¤ç‰ˆ)")
print("=" * 80)

# ==================== ğŸš© æŒ‡ä»¤ 1ï¼šæ„å»ºæ´¾ç”Ÿç‰¹å¾ ====================
print("\n" + "=" * 80)
print("ğŸš© æŒ‡ä»¤ 1ï¼šæ„å»ºæ´¾ç”Ÿç‰¹å¾")
print("=" * 80)

df = pd.read_csv(DATA_PATH)
print(f"âœ“ æ•°æ®åŠ è½½å®Œæˆ: {len(df)} è¡Œ")

# æ£€æŸ¥å¿…è¦åˆ—
required_cols = ['season', 'week', 'celebrity_name', 'weekly_total', 'estimated_fan_vote', 'is_eliminated']
if not all(col in df.columns for col in required_cols):
    raise ValueError(f"ç¼ºå¤±å¿…è¦åˆ—")

df = df.sort_values(['season', 'week', 'celebrity_name']).reset_index(drop=True)

# 1. pct_J (Judge Share)
df['pct_J'] = df.groupby(['season', 'week'])['weekly_total'].transform(lambda x: x / x.sum())

# 2. pct_V (Vote Share)
df['pct_V'] = df['estimated_fan_vote']

# 3. R_J (Normalized Judge Rank)
df['rank_J'] = df.groupby(['season', 'week'])['weekly_total'].rank(ascending=False, method='min')
df['N_t'] = df.groupby(['season', 'week'])['celebrity_name'].transform('count')
df['R_J'] = (df['N_t'] - df['rank_J']) / (df['N_t'] - 1)
df['R_J'] = df['R_J'].fillna(0)

# 4. R_V (Normalized Vote Rank)
df['rank_V'] = df.groupby(['season', 'week'])['estimated_fan_vote'].rank(ascending=False, method='min')
df['R_V'] = (df['N_t'] - df['rank_V']) / (df['N_t'] - 1)
df['R_V'] = df['R_V'].fillna(0)

print(f"âœ“ æ´¾ç”Ÿç‰¹å¾æ„å»ºå®Œæˆ")

# ==================== ğŸš© æŒ‡ä»¤ 2ï¼šå®ç°ä¸‰ç§èµ›åˆ¶è¯„åˆ†é€»è¾‘ ====================
print("\n" + "=" * 80)
print("ğŸš© æŒ‡ä»¤ 2ï¼šå®ç°ä¸‰ç§èµ›åˆ¶è¯„åˆ†é€»è¾‘")
print("=" * 80)

# è§„åˆ™ Aï¼šæ’ååˆ¶
df['S_rank'] = 0.5 * df['R_J'] + 0.5 * df['R_V']
# è§„åˆ™ Bï¼šç™¾åˆ†æ¯”åˆ¶
df['S_pct'] = 0.5 * df['pct_J'] + 0.5 * df['pct_V']

print(f"âœ“ è¯„åˆ†é€»è¾‘å®ç°å®Œæˆ")

# ==================== ğŸš© æŒ‡ä»¤ 3ï¼šæ¨¡æ‹Ÿæ·˜æ±°ä¸å‘½ä¸­ç‡è¯„ä¼° ====================
print("\n" + "=" * 80)
print("ğŸš© æŒ‡ä»¤ 3ï¼šæ¨¡æ‹Ÿæ·˜æ±°ä¸å‘½ä¸­ç‡è¯„ä¼°")
print("=" * 80)

# åˆå§‹åŒ–åˆ—
df['simulated_elim_S_rank'] = 0
df['simulated_elim_S_pct'] = 0
df['simulated_elim_judges_save'] = 0

print("æ­£åœ¨æ¨¡æ‹Ÿæ·˜æ±°ç»“æœ...")

# æ‰‹åŠ¨éå†æ¯ä¸ª (season, week) ç»„ï¼Œç¡®ä¿é€»è¾‘ç»å¯¹æ­£ç¡®
for (season, week), group_indices in df.groupby(['season', 'week']).groups.items():
    group = df.loc[group_indices].copy()
    
    # è§„åˆ™ Aï¼šæ’ååˆ¶ - S_rank æœ€ä½è€…
    min_S_rank = group['S_rank'].min()
    # å¤„ç†å¹¶åˆ—ï¼šå–ç¬¬ä¸€ä¸ª
    min_idx_rank = group[group['S_rank'] == min_S_rank].index[0]
    df.loc[min_idx_rank, 'simulated_elim_S_rank'] = 1
    
    # è§„åˆ™ Bï¼šç™¾åˆ†æ¯”åˆ¶ - S_pct æœ€ä½è€…
    min_S_pct = group['S_pct'].min()
    min_idx_pct = group[group['S_pct'] == min_S_pct].index[0]
    df.loc[min_idx_pct, 'simulated_elim_S_pct'] = 1
    
    # è§„åˆ™ Cï¼šè¯„å§”æ‹¯æ•‘åˆ¶
    if len(group) >= 2:
        # æŒ‰ S_pct æ’åºï¼Œæ‰¾å‡ºå€’æ•°ä¸¤å
        bottom_2 = group.sort_values('S_pct').head(2)
        # åœ¨å€’æ•°ä¸¤åä¸­ï¼Œè¯„å§”åˆ†(weekly_total)è¾ƒä½è€…è¢«æ·˜æ±°
        eliminated_idx = bottom_2.sort_values('weekly_total').index[0]
        df.loc[eliminated_idx, 'simulated_elim_judges_save'] = 1
    elif len(group) == 1:
        df.loc[group.index[0], 'simulated_elim_judges_save'] = 1

# è®¡ç®—å‘½ä¸­ç‡å‡½æ•°
def get_hit_rate(col):
    hits = 0
    total = 0
    for (season, week), group in df.groupby(['season', 'week']):
        if group['is_eliminated'].sum() == 0: continue # è·³è¿‡æ— æ·˜æ±°å‘¨
        
        true_elim = group[group['is_eliminated'] == 1]['celebrity_name'].values
        sim_elim = group[group[col] == 1]['celebrity_name'].values
        
        if len(sim_elim) > 0 and sim_elim[0] in true_elim:
            hits += 1
        total += 1
    return hits, total

hits_rank, total_rank = get_hit_rate('simulated_elim_S_rank')
hits_pct, total_pct = get_hit_rate('simulated_elim_S_pct')
hits_save, total_save = get_hit_rate('simulated_elim_judges_save')

print(f"\nğŸ“Š å‘½ä¸­ç‡è¯„ä¼°ç»“æœ:")
print(f"è§„åˆ™ A (æ’ååˆ¶):      {hits_rank/total_rank:.2%} ({hits_rank}/{total_rank})")
print(f"è§„åˆ™ B (ç™¾åˆ†æ¯”åˆ¶):    {hits_pct/total_pct:.2%} ({hits_pct}/{total_pct})")
print(f"è§„åˆ™ C (è¯„å§”æ‹¯æ•‘åˆ¶):  {hits_save/total_save:.2%} ({hits_save}/{total_save})")

df.to_csv(OUTPUT_DIR + 'mechanism_simulation_results.csv', index=False)

# ==================== ğŸš© æŒ‡ä»¤ 4ï¼šäº‰è®®æ¡ˆä¾‹æå– (å·²ä¿®å¤åå­—åŒ¹é…) ====================
print("\n" + "=" * 80)
print("ğŸš© æŒ‡ä»¤ 4ï¼šäº‰è®®æ¡ˆä¾‹æå– (Fixed Name Matching)")
print("=" * 80)

# ğŸ”§ ä¿®å¤ï¼šä½¿ç”¨æ•°æ®é›†ä¸­å­˜åœ¨çš„çŸ­åå­—
controversial_cases = [
    {'season': 27, 'celebrity': 'Bobby', 'description': 'Season 27 å† å†›äº‰è®®'},
    {'season': 32, 'celebrity': 'Xochitl', 'description': 'Season 32 å† å†›'},
    {'season': 32, 'celebrity': 'Jason', 'description': 'Season 32 äºšå†›äº‰è®®'},
]

for case in controversial_cases:
    season = case['season']
    name = case['celebrity']
    print(f"\nğŸ“Œ {case['description']}: {name}")
    
    # æå–æ•°æ®
    df_case = df[(df['season'] == season) & (df['celebrity_name'] == name)].copy()
    
    if len(df_case) == 0:
        print(f"âš ï¸ æœªæ‰¾åˆ°é€‰æ‰‹: {name}")
        continue

    # è®¡ç®—è¯¥é€‰æ‰‹åœ¨æ¯å‘¨çš„æ’å (Rank within the week)
    df_case['week_rank_S_rank'] = df_case.apply(
        lambda row: df[(df['season'] == row['season']) & (df['week'] == row['week'])]['S_rank'].rank(ascending=False, method='min')[row.name], axis=1
    )
    
    cols = ['week', 'weekly_total', 'rank_J', 'estimated_fan_vote', 'rank_V', 'S_rank', 'week_rank_S_rank', 'is_eliminated']
    print(df_case[cols].to_string(index=False))
    
    print(f"å¹³å‡è¯„å§”æ’å: {df_case['rank_J'].mean():.2f} | å¹³å‡ç²‰ä¸æ’å: {df_case['rank_V'].mean():.2f}")

# ==================== ğŸ”¥ é™„åŠ ï¼šé€†å‘å·¥ç¨‹ (Reverse Engineering) ====================
print("\n" + "=" * 80)
print("ğŸ”¥ é™„åŠ åˆ†æï¼šé€†å‘å·¥ç¨‹ (å¯»æ‰¾æ¶ˆå¤±çš„é€‰ç¥¨)")
print("=" * 80)

# 1. Bobby Bones æ¡ˆ
s27_final = df[(df['season'] == 27) & (df['week'] == df[df['season'] == 27]['week'].max())]
milo = s27_final[s27_final['celebrity_name'] == 'Milo'].iloc[0]
bobby = s27_final[s27_final['celebrity_name'] == 'Bobby'].iloc[0]

# Bobby éœ€è¦å¤šå°‘ç¥¨æ‰èƒ½èµ¢ Milo? (å‡è®¾ Milo åˆ†æ•°ä¸å˜)
# S_pct_Bobby > S_pct_Milo => 0.5*J + 0.5*V > Milo_Score => V > 2*Milo_Score - J
bobby_required_vote = 2 * milo['S_pct'] - bobby['pct_J']

print(f"\n[Season 27] Bobby Bones å¤ºå† é˜ˆå€¼åˆ†æ:")
print(f"  - çœŸå®å¯¹æ‰‹ (Milo) æ€»åˆ†: {milo['S_pct']:.4f}")
print(f"  - Bobby å½“å‰è¯„å§”åˆ†å æ¯”: {bobby['pct_J']:.4f}")
print(f"  - æ¨¡å‹é¢„æµ‹ç²‰ä¸å¾—ç¥¨ç‡:   {bobby['estimated_fan_vote']:.2%} (åŸºäºæœç´¢çƒ­åº¦)")
print(f"  - å®é™…å¤ºå† æ‰€éœ€å¾—ç¥¨ç‡:   {bobby_required_vote:.2%} (é€†å‘æ¨å¯¼)")
print(f"  - ç»“è®º: Bobby çš„ç²‰ä¸åŠ¨å‘˜åŠ›æ˜¯æ¨¡å‹é¢„æµ‹å€¼çš„ {bobby_required_vote/bobby['estimated_fan_vote']:.1f} å€ï¼")

# 2. Xochitl Gomez æ¡ˆ
s32_final = df[(df['season'] == 32) & (df['week'] == df[df['season'] == 32]['week'].max())]
ariana = s32_final[s32_final['celebrity_name'] == 'Ariana'].iloc[0]
xochitl = s32_final[s32_final['celebrity_name'] == 'Xochitl'].iloc[0]

xochitl_required_vote = 2 * ariana['S_pct'] - xochitl['pct_J']

print(f"\n[Season 32] Xochitl Gomez å¤ºå† é˜ˆå€¼åˆ†æ:")
print(f"  - å‡æƒ³æ•Œ (Ariana) æ€»åˆ†: {ariana['S_pct']:.4f}")
print(f"  - æ¨¡å‹é¢„æµ‹ç²‰ä¸å¾—ç¥¨ç‡:   {xochitl['estimated_fan_vote']:.2%}")
print(f"  - å®é™…å¤ºå† æ‰€éœ€å¾—ç¥¨ç‡:   {xochitl_required_vote:.2%}")
print(f"  - ç»“è®º: Xochitl éœ€è¦æ¯”é¢„æµ‹å€¼é«˜å‡º {xochitl_required_vote - xochitl['estimated_fan_vote']:.1%} çš„é€‰ç¥¨æ‰èƒ½èµ¢ã€‚")

print("\n" + "=" * 80)
print("âœ… è¿è¡Œç»“æŸ")
```

```python
#Task 2 + Task 3
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# ==================== è·¯å¾„é…ç½® ====================
RESULT_PATH = './mechanism_simulation_results.csv'
RAW_DATA_PATH = './2026_MCM_Problem_C_Data_plus_follers.xlsx'

print("=" * 80)
print("ğŸš€ DWTS æœ€ç»ˆä¿®å¤ç‰ˆï¼šäº‰è®®å¤ç›˜ & å› ç´ åˆ†æ")
print("=" * 80)

# ==================== 1. æ•°æ®åŠ è½½ä¸æ™ºèƒ½åˆå¹¶ ====================
print("ğŸ”„ æ­£åœ¨åŠ è½½å¹¶å¯¹é½æ•°æ®...")

# åŠ è½½ç»“æœè¡¨ (é•¿è¡¨ï¼šæ¯å‘¨ä¸€è¡Œ)
df_res = pd.read_csv(RESULT_PATH)
# åŠ è½½åŸå§‹è¡¨ (å®½è¡¨ï¼šæ¯äººä¸€è¡Œ)
df_raw = pd.read_excel(RAW_DATA_PATH)

# æ¸…æ´—åŸå§‹è¡¨åˆ—å
df_raw.columns = [c.strip().lower().replace(' ', '_') for c in df_raw.columns]

# è¯†åˆ«å…³é”®åˆ—
ind_col = next((c for c in df_raw.columns if 'industry' in c), 'celebrity_industry')
age_col = next((c for c in df_raw.columns if 'age' in c), 'celebrity_age_during_season')
partner_col = next((c for c in df_raw.columns if 'partner' in c), 'ballroom_partner')

print(f"   - åŸå§‹æ•°æ®å¤§å°: {df_raw.shape}")
print(f"   - ç»“æœæ•°æ®å¤§å°: {df_res.shape}")

# --- æ„å»ºæ˜ å°„å­—å…¸ (Season + Name -> Info) ---
# å› ä¸º df_res æ˜¯çŸ­å (Bobby)ï¼Œdf_raw æ˜¯é•¿å (Bobby Bones)ï¼Œæˆ‘ä»¬éœ€è¦æ¨¡ç³ŠåŒ¹é…
info_map = {} # Key: (season, short_name), Value: {age, industry, partner}

for season, group_raw in df_raw.groupby('season'):
    # è·å–è¯¥èµ›å­£ç»“æœè¡¨ä¸­çš„æ‰€æœ‰åå­—
    res_names = df_res[df_res['season'] == season]['celebrity_name'].unique()
    
    for short_name in res_names:
        # åœ¨åŸå§‹æ•°æ®ä¸­å¯»æ‰¾åŒ…å«è¿™ä¸ªçŸ­åçš„è¡Œ
        match = group_raw[group_raw['celebrity_name'].str.contains(short_name, case=False, na=False)]
        
        if len(match) > 0:
            # æ‰¾åˆ°äº†ï¼æå–ç¬¬ä¸€æ¡åŒ¹é…
            row = match.iloc[0]
            info_map[(season, short_name)] = {
                'age': row[age_col],
                'industry': row[ind_col],
                'partner': row[partner_col]
            }

# å°†æ˜ å°„åº”ç”¨åˆ°ä¸»è¡¨
print("   - æ­£åœ¨åˆå¹¶äººå£ç»Ÿè®¡ç‰¹å¾...")
def get_info(row, key):
    k = (row['season'], row['celebrity_name'])
    return info_map.get(k, {}).get(key, np.nan)

df_res['age'] = df_res.apply(lambda x: get_info(x, 'age'), axis=1)
df_res['industry'] = df_res.apply(lambda x: get_info(x, 'industry'), axis=1)
df_res['partner'] = df_res.apply(lambda x: get_info(x, 'partner'), axis=1)

print(f"âœ“ æ•°æ®åˆå¹¶å®Œæˆã€‚æˆåŠŸåŒ¹é…è¡Œæ•°: {df_res['industry'].notna().sum()}/{len(df_res)}")

# ==================== 2. Task 2: å†å²äº‰è®®å¤ç›˜ (ä¿®å¤æ ¼å¼é”™è¯¯) ====================
print("\n" + "=" * 80)
print("ğŸ”¥ Task 2: å†å²äº‰è®®å¤ç›˜ (Jerry Rice, Billy Ray, Bristol Palin)")
print("=" * 80)

targets = [
    {'season': 2, 'name': 'Jerry', 'full_name': 'Jerry Rice'},
    {'season': 4, 'name': 'Billy', 'full_name': 'Billy Ray Cyrus'},
    {'season': 11, 'name': 'Bristol', 'full_name': 'Bristol Palin'}
]

for target in targets:
    s = target['season']
    n = target['name']
    print(f"\nğŸ“Œ Case Study: {target['full_name']} (Season {s})")
    
    mask = (df_res['season'] == s) & (df_res['celebrity_name'].str.contains(n, case=False, na=False))
    df_case = df_res[mask].copy()
    
    if df_case.empty:
        print(f"   âš ï¸ æœªæ‰¾åˆ°æ•°æ®: {n}")
        continue

    print(f"   {'Week':<5} {'Judge_Rank':<12} {'Fan_Rank':<10} {'S_rank':<10} {'S_pct':<10} {'Diff'}")
    print(f"   {'-'*70}")
    
    diff_sum = 0
    for idx, row in df_case.iterrows():
        week = row['week']
        # è·å–å½“å‘¨æ•°æ®è®¡ç®—æ’å
        week_df = df_res[(df_res['season'] == s) & (df_res['week'] == week)]
        
        # é‡æ–°è®¡ç®—å½“å‘¨æ’å
        rank_in_rank = week_df['S_rank'].rank(ascending=False, method='min')
        rank_in_pct = week_df['S_pct'].rank(ascending=False, method='min')
        
        # æ‰¾åˆ°å½“å‰é€‰æ‰‹çš„æ’å
        my_rank_rank = rank_in_rank.loc[idx]
        my_rank_pct = rank_in_pct.loc[idx]
        
        # å·®å¼‚è®¡ç®—
        diff = my_rank_rank - my_rank_pct
        diff_sum += diff
        
        # ä¿®å¤ï¼šä½¿ç”¨ .1f æ ¼å¼åŒ–æµ®ç‚¹æ•°ï¼Œè€Œä¸æ˜¯ d
        print(f"   {week:<5} {row['rank_J']:<12.1f} {row['rank_V']:<10.1f} {my_rank_rank:<10.0f} {my_rank_pct:<10.0f} {diff:+.1f}")

    print(f"   ğŸ’¡ ç»“è®º:")
    if diff_sum < -1:
        print(f"      ğŸ‘‰ Rank Rule (æ’ååˆ¶) æ•‘äº†ä»–ï¼å¦‚æœç”¨ç™¾åˆ†æ¯”åˆ¶ï¼Œä»–çš„æ’åä¼šæ˜¾è‘—ä¸‹é™ (Diff: {diff_sum:.1f})ã€‚")
        print(f"      è¿™è§£é‡Šäº†ä¸ºä»€ä¹ˆ Season 2 ä¹‹åèŠ‚ç›®ç»„å°è¯•å¼•å…¥ç™¾åˆ†æ¯”åˆ¶ã€‚")
    elif diff_sum > 1:
        print(f"      ğŸ‘‰ Percentage Rule (ç™¾åˆ†æ¯”åˆ¶) å¯¹ä»–æ›´æœ‰åˆ©ã€‚")
    else:
        print(f"      ğŸ‘‰ ä¸¤ç§èµ›åˆ¶ä¸‹å‘½è¿åŸºæœ¬ä¸€è‡´ (Diff: {diff_sum:.1f})ã€‚")

# ==================== 3. Task 3: å› ç´ åˆ†æ (Factor Analysis) ====================
print("\n" + "=" * 80)
print("ğŸ§¬ Task 3: å½±å“å› ç´ åˆ†æ (Pro Dancers, Industry, Age)")
print("=" * 80)

# å‡†å¤‡å›å½’æ•°æ®
model_df = df_res.dropna(subset=['weekly_total', 'age', 'industry']).copy()

if len(model_df) > 100:
    # 1. è¡Œä¸šç¼–ç  (Top 5)
    top_ind = model_df['industry'].value_counts().nlargest(5).index
    model_df['ind_group'] = model_df['industry'].apply(lambda x: x if x in top_ind else 'Other')
    
    # 2. èˆä¼´èƒ½åŠ›é‡åŒ– (Partner Skill)
    # è®¡ç®—æ¯ä¸ªèˆä¼´çš„å†å²å¹³å‡åˆ†
    partner_stats = model_df.groupby('partner')['weekly_total'].mean()
    model_df['partner_skill'] = model_df['partner'].map(partner_stats)
    
    # 3. æ ‡å‡†åŒ–
    model_df['age_z'] = (model_df['age'] - model_df['age'].mean()) / model_df['age'].std()
    model_df['partner_z'] = (model_df['partner_skill'] - model_df['partner_skill'].mean()) / model_df['partner_skill'].std()
    
    print(f"   - æ ·æœ¬é‡: {len(model_df)}")
    print(f"   - åŒ…å«è¡Œä¸š: {list(top_ind)}")
    
    # --- å›å½’æ¨¡å‹ ---
    print("\nğŸ“Š å›å½’åˆ†æç»“æœ (Dependent Var: Judges Score)")
    
    X = pd.get_dummies(model_df[['age_z', 'ind_group', 'partner_z']], drop_first=True)
    X = sm.add_constant(X)
    y = model_df['weekly_total']
    
    try:
        model = sm.OLS(y, X.astype(float)).fit()
        
        # æå–æ˜¾è‘—å› å­
        summary = model.summary2().tables[1]
        sig = summary[summary['P>|t|'] < 0.05].sort_values('Coef.', ascending=False)
        print(sig[['Coef.', 'P>|t|']])
        
        # ç»˜å›¾
        plt.figure(figsize=(10, 6))
        coefs = sig['Coef.'].drop('const', errors='ignore')
        if len(coefs) > 0:
            coefs.plot(kind='barh', color='#2C3E50')
            plt.title('Significant Factors Affecting Scores (p < 0.05)')
            plt.xlabel('Impact on Score')
            plt.grid(axis='x', linestyle='--', alpha=0.5)
            plt.tight_layout()
            plt.savefig('factor_analysis_final.png')
            print("\nâœ… å›¾è¡¨å·²ä¿å­˜: factor_analysis_final.png")
        else:
            print("âš ï¸ æœªå‘ç°æ˜¾è‘—å› å­")
            
    except Exception as e:
        print(f"âŒ å›å½’åˆ†æå‡ºé”™: {e}")
else:
    print("âš ï¸ æœ‰æ•ˆæ ·æœ¬ä¸è¶³ï¼Œæ— æ³•è¿›è¡Œå› ç´ åˆ†æã€‚")

print("\nâœ… è¿è¡Œç»“æŸã€‚")
```

```python
#Task 4
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ==================== é…ç½® ====================
RESULT_PATH = './mechanism_simulation_results.csv'
df = pd.read_csv(RESULT_PATH)

print("\n" + "=" * 80)
print("âš–ï¸ Task 4 ä¿®æ­£: åŠ æƒ SIS (Weighted Standardized Impact Score)")
print("   ç­–ç•¥: 65% è¯„å§”æƒé‡ / 35% ç²‰ä¸æƒé‡")
print("=" * 80)

def calculate_weighted_sis(df_in, w_judge=0.65):
    df_out = df_in.copy()
    w_fan = 1.0 - w_judge
    
    groups = df_out.groupby(['season', 'week'])
    
    # 1. è®¡ç®— Z-Scores
    df_out['Z_Judge'] = groups['weekly_total'].transform(
        lambda x: (x - x.mean()) / (x.std(ddof=0) + 1e-6)
    )
    df_out['Z_Fan'] = groups['estimated_fan_vote'].transform(
        lambda x: (x - x.mean()) / (x.std(ddof=0) + 1e-6)
    )
    
    # 2. åŠ æƒæ€»åˆ† (æ ¸å¿ƒä¿®æ­£ç‚¹)
    df_out['SIS_Score'] = (w_judge * df_out['Z_Judge']) + (w_fan * df_out['Z_Fan'])
    
    # 3. æ’å
    df_out['SIS_Rank'] = groups['SIS_Score'].rank(ascending=False, method='min')
    
    return df_out

# è¿è¡Œè®¡ç®—
df_sis = calculate_weighted_sis(df, w_judge=0.65)

# ==================== éªŒè¯ç»“æœ ====================

# 1. Bobby Bones (S27) - åº”è¯¥è¢«å‹åˆ¶
bobby = df_sis[(df_sis['season'] == 27) & (df_sis['celebrity_name'].str.contains('Bobby'))].sort_values('week')
print(f"\nğŸ§ª 1. Bobby Bones (S27) - æ¯’ç˜¤å‹åˆ¶æµ‹è¯•")
print(f"   {'Week':<5} {'Old_Rank':<10} {'New_Rank':<10} {'Z_Judge (65%)':<15} {'Z_Fan (35%)':<15} {'Score'}")
print("-" * 70)
for _, row in bobby.iterrows():
    print(f"   {row['week']:<5} {row['S_rank']:<10.0f} {row['SIS_Rank']:<10.0f} {row['Z_Judge']:<15.2f} {row['Z_Fan']:<15.2f} {row['SIS_Score']:.2f}")

# 2. Xochitl (S32) - åº”è¯¥è¢«ä¿æŠ¤
xochitl = df_sis[(df_sis['season'] == 32) & (df_sis['celebrity_name'].str.contains('Xochitl'))].sort_values('week')
print(f"\nğŸ›¡ï¸ 2. Xochitl (S32) - å®åŠ›æ´¾ä¿æŠ¤æµ‹è¯•")
print(f"   {'Week':<5} {'Old_Rank':<10} {'New_Rank':<10} {'Z_Judge (65%)':<15} {'Z_Fan (35%)':<15} {'Score'}")
print("-" * 70)
avg_old = xochitl['S_rank'].mean()
avg_new = xochitl['SIS_Rank'].mean()
for _, row in xochitl.iterrows():
    status = "ğŸ‘‘" if row['SIS_Rank'] <= 3 else " "
    print(f"   {row['week']:<5} {row['S_rank']:<10.0f} {row['SIS_Rank']:<10.0f} {row['Z_Judge']:<15.2f} {row['Z_Fan']:<15.2f} {row['SIS_Score']:.2f} {status}")

print(f"\nğŸ“Š Xochitl å¹³å‡æ’å: æ—§åˆ¶ {avg_old:.2f} -> æ–°åˆ¶ {avg_new:.2f}")
if avg_new < 2.5:
    print("âœ… ç»“è®º: å®Œç¾ï¼Xochitl ç¨³å±…å‰ 2.5 åï¼Œæ—¢èµ¢å¾—äº†æ¯”èµ›ï¼Œåˆä¿ç•™äº†æ‚¬å¿µã€‚")
else:
    print("âš ï¸ ç»“è®º: ä»ç„¶æœ‰æ³¢åŠ¨ï¼Œä½†æ¯” 50/50 ç‰ˆæœ¬å¥½å¾ˆå¤šã€‚")

# ==================== ç»˜å›¾ ====================
plt.figure(figsize=(12, 6))
plt.plot(bobby['week'], bobby['SIS_Rank'], 'r-s', label='Bobby (Weighted SIS)', linewidth=3)
plt.plot(xochitl['week'], xochitl['SIS_Rank'], 'g-^', label='Xochitl (Weighted SIS)', linewidth=3)
# æ·»åŠ ä¸€æ¡å‚è€ƒçº¿
plt.axhline(y=1, color='gold', linestyle='--', alpha=0.5, label='First Place')

plt.gca().invert_yaxis()
plt.title('The "Weighted SIS" (65/35) System Performance', fontsize=16)
plt.ylabel('Rank (Lower is Better)')
plt.xlabel('Week')
plt.legend()
plt.grid(True, alpha=0.3)
plt.savefig('task4_weighted_sis.png')
print("\nâœ… æœ€ç»ˆæ–¹æ¡ˆå›¾è¡¨å·²ä¿å­˜: task4_weighted_sis.png")
```
